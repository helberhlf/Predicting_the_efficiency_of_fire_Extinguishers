### 1.2. Sobre Dataset

O conjunto de dados foi obtido como resultado dos testes de extino de quatro chamas  de combustveis diferentes com um sistema de extino de ondas sonoras. O sistema de extino  de incndio por ondas sonoras consiste em 4 subwoofers com uma potncia total de 4.000 Watts.  Existem dois amplificadores que permitem que o som chegue a esses subwoofers amplificado. A fonte de alimentao que alimenta o sistema e o circuito do filtro garantindo que  as frequncias de som sejam transmitidas adequadamente para o sistema est localizada dentro  da unidade de controle. 

Enquanto o computador  usado como fonte de frequncia, o  anemmetro foi usado para medir o fluxo de ar resultante das ondas sonoras durante a fase de  extino da chama e um decibelmetro para medir a intensidade do som. Um termmetro  infravermelho foi utilizado para medir a temperatura da chama e da lata de combustvel, e uma  cmera  instalada para detectar o tempo de extino da chama. Um total de 17.442 testes foram  realizados com esta configurao experimental. Os experimentos foram planejados da seguinte  forma:

1. Trs combustveis lquidos diferentes e combustvel GLP foram usados para criar a chama.
2. 5 tamanhos diferentes de latas de combustvel lquido so usados para atingir diferentes tamanhos de chamas.

3. O ajuste de meio e cheio de gs  usado para combustvel GLP.

4. Durante a realizao de cada experimento, o recipiente de combustvel, a 10 cm de distncia, foi movido para frente at 190 cm aumentando a distncia em 10 cm a cada vez.

5. Junto com o recipiente de combustvel, o anemmetro e o decibelmetro foram movidos para frente nas mesmas dimenses.

6. Experimentos de extino de incndio foram conduzidos com 54 ondas sonoras de frequncias diferentes em cada distncia e tamanho de chama.

O conjunto de dados inclui as  caractersticas do tamanho do recipiente de combustvel representando o tamanho da chama,  tipo de combustvel, frequncia, decibis, distncia, fluxo de ar e extino da chama. Assim, 6 
recursos de entrada e 1 recurso de sada sero usados no modelo que voc vai construir.  A coluna de status (extino de chama ou no extino da chama) pode ser prevista  usando os seis recursos de entrada no conjunto de dados. Os recursos de status e combustvel  so categricos, enquanto outros recursos so numricos.

### 2. Aprendizagem Supervisionada e No Supervisionada


Os termos supervisionado e no supervisionado foram herdados do campo de Aprendizado de Mquina, em ingls Machine Learning - ML ou AM. Metaforicamente, um professor "supervisiona o aluno fornecendo, cuidadosamente, informaes alvo, junto com um conjunto de exemplos. Uma tarefa de aprendizado no supervisionada pode implicar no mesmo conjunto de exemplos, mas no incluiria as informaes alvo (destino). O aluno no receberia informao sobre o objetivo do aprendizado, mas ficaria livre para tirar as prprias concluses sobre o que os exemplos tm em comum **(FAWCETT and PROVOST, 2016)**.

Os sistemas de Aprendizado de Mquina podem ser classificados de acordo com a quantidade e o tipo de superviso que recebem durante o treinamento. Existem quatro categorias principais de aprendizado: supervisionado, no supervisionado, semi-supervisionado e por reforo **(GRON, 2021)**
    
    
 ### 2.1. Modelos, Induo e Previso

Em geral, um modelo  uma representao simplificada da realidade criada para servir a um propsito.Assim,ele se baseia-se em algumas simplificaes de suposies sobre o que  importante e o que no  especficos ou s vezes com base em restries de informaes ou tratabilidade. Por exemplo, um mapa  um modelo do mundo fsico.Ele abstrai uma enorme quantidade de informaes que o cartgrafo considera irrelevantes para sua finalidade. Ele preserva e, s vezes, simplifica ainda mais, as informaes relevantes.**(FAWCETT and PROVOST, 2016)**.


Em **Data Science**, um modelo preditivo  uma frmula para estimar o valor **desconhecido de interesse**: **o alvo**. A frmula pode ser matemtica ou pode ser uma declarao lgica, como uma regra. Muitas vezes,  um hbrido dos dois. Dada nossa diviso de minerao supervisionada de dados em **classificao e regresso**, vamos considerar modelos de classificao (modelos que prev a estimativa de probabilidade de classe) **(FAWCETT and PROVOST, 2016)**.

<font color='red'>Lembre-se de que os modelos preditivos se concentram em estimar o valor da varivel alvo de interesse.</font>

### 2.2. Termo-chave: previso
No uso comum, **previso significa antecipar um evento futuro**. Em **Data Science**, de forma mais geral,previso significa **estimar um valor desconhecido**. Esse valor poderia ser algo no futuro (no uso comum, previso verdadeira), mas tambm pode ser algo no presente ou no passado. De fato, como a minerao de dados geralmente lida com dados histricos, os modelos costumam ser construdos e testados por meio do **uso de eventos do passado**. Modelos preditivos para pontuao de crdito estimam a probabilidade de um potencial cliente ser inadimplente. Modelos preditivos para filtragem de spam estimam se determinado e-mail  spam. Modelos preditivos para a deteco de fraude julgam se uma conta foi defraudada. O segredo  que o modelo seja destinado para **estimar  um valor desconhecido**.**(FAWCETT and PROVOST, 2016)**.


### 2.2. Termo-chave: Induo e Deduo

A induo pode ser comparada com a **deduo**.Porm, comea com regras gerais e fatos especficos, e cria outros fatos especficos a partir deles.A utilizao dos nossos modelos pode ser considerada um procedimento de deduo (probabilstica).**(FAWCETT and PROVOST, 2016)**.

A criao de modelos a partir de dados  conhecida como modelo de induo.A induo  um termo da filosofia que se refere a generalizao a partir de casos especficos at regras gerais (ou leis ou verdades). Nossos modelos so regras gerais no sentido estatstico (eles normalmente no se mantm 100% das vezes; com frequncia, nem chegam perto), e o procedimento que cria o modelo a partir de dados  chamado algoritmo de induo ou aprendiz. **(FAWCETT and PROVOST, 2016)**.

### 5.1. Create Funes e Pipelines auxiliares

#### 5.4.1.2. Codificando Recursos Categricos Nominais

- Problema 

Podemos ter recursos (variveis) com classes nominais que no possui ordenao intrnseca (Blue, Red, Green ).


- Soluo 

Codifique o recurso usando o LabelBinarizer do scikit-learn:


Nesse sentido, de acordo (GRON , 2021) refina que um dos problemas com essa representao  que os algoritmos de **Machine Learning** assumem que dois valores prximos so mais semelhantes que dois valores
distantes. Isso pode ser bom em alguns casos (por exemplo, para categorias ordenadas como ruim, mdio, bom e excelente), mas obviamente no  o caso da coluna **fuel** (por exemplo, as categorias 0 e 4 so
mais semelhantes que as categorias 0 e 1). Para corrigir esse problema, uma soluo comum  criar um atributo binrio (LabelBinarizer) por categoria: um atributo igual a 1 quando a categoria for <gasoline (e diferente de 0) e outro atributo igual a 1 quando a categoria for kerosene (e diferente de 0) e assim por diante. Isso se chama **codificao one-hot [one-hot encoding ou ainda codificao distribuda]**, porque apenas um atributo ser igual a 1, enquanto os outros sero 0. Os atributos novos se chamam de atributos **falsos [dummy]**. 

Segundo (ALBON , 2018 ) ressaltar que a estratgia adequada  criar um recurso binrio para cada classe no recurso original. Isso geralmente  chamado de codificao one-hot (na literatura de aprendizado de mquina) ou dummy (na literatura estatstica e de pesquisa). O recurso da nossa soluo era um vetor contendo quatro classes (ou seja, Gasoline, Kerosene, LPG e  Thinner ). Na codificao **one-hot**, cada classe se torna seu prprio recurso com **1s** quando a classe aparece e **0s** caso contrrio. Como nosso recurso tem quatro classes, a **codificao one-hot** retornou quatro recursos binrios (um para cada classe). Usando a codificao **one-hot**, podemos capturar a associao de uma observao em uma classe, preservando a noo de que a classe no possui qualquer tipo de hierarquia. Em sntese, vale a pena notar que muitas vezes  recomendado que aps uma
codificando um recurso,**descartamos um dos recursos codificados one-hot na matriz resultante para evitar dependncia linear.**


Como regra geral, **<font color="red">se o nmero de categorias for inferior a 10</font>**, a **codificao one-hot**  usualmente a melhor opo (mas seu rendimento pode variar!)

```python
# construindo um pipeline para pr-processar os atributos numricos:
num_pipeline = Pipeline([
    ('min_max', MinMaxScaler()),
])

# Criando um  pipeline  para Transformao One Hot Encoding
cat_pipeline = Pipeline([
    ('ohe',OneHotEncoder(handle_unknown='ignore',sparse=False)) #Defina o handle_unknown='ignore' do OneHotEncoder para evitar avisos
],
)
# criando Pipelilne para transformao Label Encoder
le_pipeline = Pipeline([
    ('le', CustomLabelEncoder()), 
])
```


**Extraindo nomes de coluna do ColumnTransformer**


<font color="red">No Scikit-learn a funo **ColumnTransformer**  uma tima ferramenta para pr-processamento de dados, mas retorna um array numpy sem nomes da colunas. Seu mtodo get_feature_names() falhar se pelo menos um transformador no criar novas colunas. A seguinte funo auxiliar  rpida de ser construda em torno do get_feature_names() mtodo do ColumnTransformer, que pode ser encontrado aqui: https://github.com/scikit-learn/scikit-learn/blob/fd237278e895b42abe8d8d09105cbb82dc2cbba7/sklearn/compose/_column_transformer.py#L345

A funo percorre as etapas do ColumnTransformere retorna os nomes das colunas de entrada quando o transformador no fornece um get_feature_names() mtodo.Porm, para os pipelines, ele percorre as etapas do pipeline e retornar as colunas de sada do pipeline ou as colunas de entrada, se o pipeline no criar novas colunas.

*References*

<a href ="https://johaupt.github.io/blog/columnTransformer_feature_names.html" target="__blank">Extracting Column Names from the ColumnTransformer</a>

<a href="https://github.com/scikit-learn/scikit-learn/issues/12525">Cannot get feature names after ColumnTransformer #12525</a>

Com base nos conceitos discutido anteriormente, a funo adiante  pouco complicada, ela basicamente retorna o nomes das variveis dos dados transformados ao final da aplicao.O cdigo a seguir  sobre a <a href="https://github.com/scikit-learn/scikit-learn/issues/12525">Cannot get feature names after ColumnTransformer #12525</a>

```python
from skimpy import clean_columns
from sklearn.utils.validation import check_is_fitted

def get_column_names_from_ColumnTransformer(column_transformer, clean_column_names=True, verbose=True):  

    """
    Reference: Kyle Gilde: https://github.com/kylegilde/Kaggle-Notebooks/blob/master/Extracting-and-Plotting-Scikit-Feature-Names-and-Importances/feature_importance.py
    Description: Get the column names from the a ColumnTransformer containing transformers & pipelines
    Parameters
    ----------
    verbose: Bool indicating whether to print summaries. Default set to True.
    Returns
    -------
    a list of the correct feature names
    Note:
    If the ColumnTransformer contains Pipelines and if one of the transformers in the Pipeline is adding completely new columns,
    it must come last in the pipeline. For example, OneHotEncoder, MissingIndicator & SimpleImputer(add_indicator=True) add columns
    to the dataset that didn't exist before, so there should come last in the Pipeline.
    Inspiration: https://github.com/scikit-learn/scikit-learn/issues/12525
    """

    assert isinstance(column_transformer, ColumnTransformer), "Input isn't a ColumnTransformer"
    
    check_is_fitted(column_transformer)

    new_feature_names, transformer_list = [], []

    for i, transformer_item in enumerate(column_transformer.transformers_): 
        transformer_name, transformer, orig_feature_names = transformer_item
        orig_feature_names = list(orig_feature_names)

        if len(orig_feature_names) == 0:
            continue

        if verbose: 
            print(f"\n\n{i}.Transformer/Pipeline: {transformer_name} {transformer.__class__.__name__}\n")
            print(f"\tn_orig_feature_names:{len(orig_feature_names)}")

        if transformer == 'drop':
            continue

        if isinstance(transformer, Pipeline):
            # if pipeline, get the last transformer in the Pipeline
            transformer = transformer.steps[-1][1]

        if hasattr(transformer, 'get_feature_names_out'):
            if 'input_features' in transformer.get_feature_names_out.__code__.co_varnames:
                names = list(transformer.get_feature_names_out(orig_feature_names))
            else:
                names = list(transformer.get_feature_names_out())
        elif hasattr(transformer, 'get_feature_names'):
            if 'input_features' in transformer.get_feature_names.__code__.co_varnames:
                names = list(transformer.get_feature_names(orig_feature_names))
            else:
                names = list(transformer.get_feature_names())

        elif hasattr(transformer,'indicator_') and transformer.add_indicator:
            # is this transformer one of the imputers & did it call the MissingIndicator?

            missing_indicator_indices = transformer.indicator_.features_
            missing_indicators = [orig_feature_names[idx] + '_missing_flag'\
                                  for idx in missing_indicator_indices]
            names = orig_feature_names + missing_indicators

        elif hasattr(transformer,'features_'):
            # is this a MissingIndicator class? 
            missing_indicator_indices = transformer.features_
            missing_indicators = [orig_feature_names[idx] + '_missing_flag'\
                                  for idx in missing_indicator_indices]

        else:

            names = orig_feature_names

        if verbose: 
            print(f"\tn_new_features:{len(names)}")
            print(f"\tnew_features: {names}\n")

        new_feature_names.extend(names)
        transformer_list.extend([transformer_name] * len(names))

    transformer_list, column_transformer_features = transformer_list, new_feature_names

    if clean_column_names:
        new_feature_names = list(clean_columns(pd.DataFrame(columns=new_feature_names)).columns)
    
    return new_feature_names
```

### 6.1. Remoo de Outliers usando intervalo interquartil ou IQR via Boxplot

Portanto, antes de entendermos esse mtodo de remoo de valores discrepantes, primeiro precisamos entender os percentis.

### 6.1.2. O que so Percentiles

**Segundo (ANDREW e BRUCE, 2019) uma abordagem diferente para estimar disperso  com base na observao da distribuio dos dados classificados. A estatstica baseada em dados classificados (ordenados)  chamada de estatstica ordinal. A medida mais bsica  a amplitude: a diferena entre o maior e o menor nmero.  bom conhecer os valores mximos e mnimos, que so teis na identificao dos outliers, mas a amplitude  extremamente sensvel aos outliers e no  muito til como medida geral da disperso dos dados. Para evitar a sensibilidade aos outliers, podemos observar a amplitude dos dados depois da excluso dos valores de cada ponta. Formalmente, esses tipos de estimativas so baseados em diferenas entre percentis.**<font color="red"> Em um conjunto de dados, o percentil P  um valor em que ao menos P por cento dos valores assume esse valor ou menos, e ao menos (100  P) por cento dos valores assume esse valor ou mais. Por exemplo, para encontrar o 80 percentil, classifique os dados. Ento, comeando com o menor valor, avance 80% em direo ao maior valor. Note que a mediana  a mesma coisa que o 50 percentil. O percentil  basicamente o mesmo que um quantil, com quantis indexados por fraes (ento o quantil .8  igual ao 80 percentil). Uma medida comum da variabilidade  a diferena entre o 25 percentil e o 75 percentil, chamada de amplitude interquartil (ou IQR).</font>

---
Em outras palavras, um percentil indica o valor abaixo do qual uma determinada porcentagem de observaes em um grupo de observaes cai.Dessa maneira, pense em classificar o conjunto de dados contendo 100 valores e dividi-lo em 100 partes iguais, agora o valor em qualquer lugar, digamos, no 10  o nosso 10 percentil, ou seja, o valor no ndice 10 indica abaixo do qual 10% dos valores caem.

---
Depois de entender os percentis,  fcil entender o IQR e determinar os limites. Quo? podemos analisar primeiro analisar o IQR.


### 6.1.3.  Interquatile range or IQR

O IQR descreve os 50% dos valores quando classificados do menor para o maior. IQR  a diferena entre o terceiro quartil e o primeiro quartil.

**Assim,o IQR seria  grau confiana ou intervalo de confiana 25% (conhecido como Q1) dos dados esto abaixo dele e o quartil 75% (Q3)  o valor onde 25% dos dadaos esto acima desse. De fato, esse conceito de outliers  via boxplots, foi mtodo mais simples onde os estatsticos da poca conseguiria detectar um outlier de forma visual para uma varivel e formulou essa frmula na ilustrado na figura a seguir**


</br><center >Source Image: (TYAGI, 2019)</center></br>
Portanto, pesquisadores vem elaborandos  tcnicas mais sofisticadas para deteco de outliers, um ponto importante nesse meio termo  a exploso da **Data Science agredado com Big Data e Estatstica moderna**. Atualmente temos os intervalos de confiana frequentemente os mais usados **80%, 85% ,90%, 95% e 99%**


---
**<mark>Portanto, um regra prtica  aparar os 10 % iniciais e finais dos dados oferer , quase sempre, uma proteo contra outliers, exceto nos conjunto de dados menores.</mark>** **(ANDREW e BRUCE, 2019)**.


### 6.1.3. O que  quartis?

Da mesma forma percentil, um quartil corta os dados em 4 partes iguais. Portanto, o percentil 25  nosso primeiro quartil, o percentil 50  o segundo quartil, o percentil 75  nosso terceiro quartil.

Dessa forma  podemos implementa uma funo para analisar de acordo aos conceitos discutido antes, preservando 5% dos dados iniciais e finais, ou seja, com **(Q1 = 5% e Q3 = 95%)**, lembrando que devemos sempre comea a deteco com valores conservadores, assim como, o que estiver fora desse intervalo ser considerado outliers

```python

'''A funo retorna um vetor booleano: True se o elemento for um outlier. Falso, caso contrrio.
Agora, para substituir os outliers superiores e inferiores, vamos escrever outra pequena funo e aplic-la em todo o dataframe:'''
def outliers(x): 
       return np.abs(x- x.median()) > 1.5*(x.quantile(0.95)-x.quantile(0.05))
    
'''Substitua os valores outliers mais altos pelo percentil 95 e os valores outliers mais baixos pelo percentil 5'''    
def replace(x):
    out = x[outliers(x)] #  um vector
    return x.replace(to_replace = [out.min(), out.max()],
                    value = (np.percentile(x,5), np.percentile(x,95)))
```

```python
# Aplique replace() em cada coluna do dataframe
df = df.apply(replace,axis=1)
print('Dados antes da aplicao',df.shape)

# Por fim, remova as linhas que contm qualquer valores outliers pelo mtodo mais simples 
df = df[~df.apply(outliers).any(axis=1)]

print('\nDados depois da aplicao',df.shape)
```

### 7.1 Split 


Nosso objetivo no  avaliar o desempenho do modelo em nossos dados de treinamento, mas o desempenho em dados nunca vistos antes. **Por esse motivo, nosso mtodo de avaliao deve nos ajudar a entender como os modelos so capazes de fazer previses a partir de dados que nunca viram antes.**


Nesse sentido (ALBON, 2018) refina que uma estratgia pode ser adiar uma fatia de dados para teste. Isso  chamado de validao (ou reteno). Na validao, nossas observaes (recursos e alvos) so dividido em dois conjuntos, tradicionalmente chamados de conjunto de **treinamento e conjunto de teste**. Pegamos o conjuntdo de teste e o colocamos de lado, fingindo que nunca o vimos antes. Em seguida, treinamos nosso modelo usando nosso conjunto de treinamento, usando os recursos e o vetor de destino para ensinar o modelo a fazer a melhor previso. Por fim, simulamos nunca ter visto dados externos, avaliando o desempenho do nosso modelo treinado em nosso conjunto de treinamento em nosso conjunto de teste. **No entanto, a abordagem de validao tem duas grandes fraquezas. Primeiro, o desempenho do modelo pode ser altamente dependente de quais poucas observaes foram selecionadas para o conjunto de teste. Segundo, o modelo no est sendo treinado usando todos os dados disponveis e no est sendo avaliado em todos os dados disponveis.**


Uma estratgia melhor, que supera essas fraquezas,  chamada de validao cruzada k-fold (KFCV). No KFCV, dividimos os dados em k partes chamadas folds. O modelo  ento treinado usando k  1 dobras combinadas em um conjunto de treinamento e ento a ltima dobra  usada como um conjunto de teste. Repetimos isso k vezes, cada vez usando uma dobra diferente como conjunto de teste. O desempenho no modelo para cada uma das k iteraes  ento calculado para produzir uma medio geral.

Nessa mesma linha de trabalho, (ALBON, 2018) ressaltar que h trs pontos importantes a serem considerados quando estamos usando KFCV. Primeiro, KFCV assume que cada observao foi criada independente da outra (ou seja, os dados so distribudos de forma idntica independente.Se os dados forem **distribudos de forma idntica independente**,  uma boa ideia embaralhar as observaes ao atribuir as (sees/dobras). No scikit-learn podemos definir **shuffle=True** para realizar o embaralhamento.Em segundo lugar, quando estamos usando KFCV para avaliar um classificador, geralmente  benfico ter dobras contendo aproximadamente a mesma porcentagem de observaes de cada uma das diferentes classes de destino (chamadas k-fold estratificada). <font color ="#f000">Por exemplo, se nosso vetor de destino contiver gnero e 80% das observaes forem masculinas, cada dobra conter 80% observaes masculinas e 20% femininas. No Scikit-learn, podemos coordenar a validao cruzada de k-folds estratificadas substituindo a classe KFold por StratifiedKFold.</font>

Dessa forma, se analisarmos a classe target ou alvo do projeto, conforme mostra a figura a seguir, no precisamos  usar **StratifiedKFold** para este propsito, pois a classe destino est relativamente bem distribuda.




**StratifiedKFold**

Se usarmos o KFold para um conjunto de dados desbalanceado, podemos acabar  com dados de treinamento que contm poucas ou nenhuma classe minoritria.Portanto, para evitar esse problema, usamos o StratifiedKFold. **StratifiedKFold  uma verso melhorada de k-Fold que retorna folds estratificadas: cada conjunto contm aproximadamente a mesma porcentagem de amostras de cada classe de destino que o conjunto completo**.

Em outras palavras, StratifiedKFold embaralha seus dados (amostras sem reposio), depois divide os dados em n_splits partes e pronto. Agora, ele usar cada parte como um conjunto de teste. Observe que apenas e sempre embaralha os dados uma vez antes de dividir. Isso significa que cada parte preserva a mesma porcentagem de amostras de cada classe (rtulo) que os dados originais.

### 8. Deteco de anomalias usando as misturas de gaussianas


Segundo (GRON, 2021) um modelo de **mistura de gaussianas (GMM)**  um modelo probabilstico que assume que as instncias foram geradas a partir de uma mistura de vrias distribuies gaussianas cujos parmetros so desconhecidos. Todas as instncias geradas a partir de uma nica distribuio gaussiana formam um cluster que normalmente se parece com um elipsoide. Cada cluster pode ter uma forma, tamanho, densidade e orientao elipsoidal diferentes, conforme demonstrado na (Figura adiante). Ao observar uma instncia, voc sabe que ela foi gerada a partir de uma das distribuies gaussianas, mas no sabe dizer qual delas e no sabe quais so os parmetros dessas distribuies.



Oberserva-se (GRON, 2021) ressaltar que **usar um modelo de mistura de gaussianas para deteco de anomalias  bastante simples: qualquer instncia localizada em uma regio de baixa densidade pode ser considerada uma anomalia**. Voc deve definir o limiar de densidade que quer usar. Por exemplo, em uma empresa de manufatura que tenta identificar produtos defeituosos, quando a proporo desses produtos geralmente  conhecida. Digamos que seja igual a 4%. Logo, voc define o limiar de densidade como um valor de 4% resultante das instncias localizadas em reas abaixo desse limiar. Caso perceba que obtm muitos falsos positivos (ou seja, produtos perfeitamente bons e sinalizados como defeituosos), voc pode diminuir o limiar. Por outro lado, caso tenha muitos falsos negativos (ou seja, produtos defeituosos que o sistema no sinaliza como tal), poder aumentar esse limiar. Isso se chama trade-off preciso/revocao


> **Os modelos de mistura de gaussianas tentam ajustar todos os dados, incluindo os outliers; desse modo, se voc tiver muitos deles, isso influencia a viso de normalidade do modelo, e alguns outliers podem ser considerados erroneamente como normais. Se isso ocorrer, voc pode tentar ajustar o modelo uma vez, us-lo para detectar e remover os outliers mais extremos e ajustar o modelo novamente no conjunto de dados limpo. Outra abordagem  utilizar mtodos robustos de estimativa de covarincia (confira a classe EllipticEnvelope).**

Por outro lado, os valores de outros parmetros (normalmente os pesos dos ns) so aprendidos. Observa-se que (TOUSHIK) ressaltar alguns exemplos de hiperparmetros de modelo incluem: A penalidade na regularizao do Classificador de Regresso Logstica, ou seja, L1 ou L2. A taxa de aprendizado para treinar uma rede neural. Os hiperprmetros C e sigma para mquinas vetoriais de suporte.


> Por que usamos o ajuste de hiperparmetros?

Note-se,(TOUSHIK) refina a importncia do ajuste de hiperparmetros  uma parte essencial do controle do comportamento de um modelo de aprendizado de mquina. Se no ajustarmos corretamente nossos hiperparmetros, nossos parmetros de modelo estimados produziro resultados abaixo do ideal, pois no minimizaro a funo de perda. **Isso significa que nosso modelo comete mais erros.**

--- 
Dessa forma, no temos como conhecer o melhor valor para um hiperparmetro em um determinado problema. Podemos usar regras bsicas, copiar valores usados em outros problemas ou  procurar o melhor valor por tentativa e erro. Quando um algoritmo de aprendizagem de mquina  ajustado para um problema  especfico, usando por exemplo **grid search ou random search**, voc est ajustando os hiperparmetros do modelo para descobrir os parmetros do modelo que resultam nas previses mais precisas.Assim acresce-se que, quando se trata de otimizar um modelo, voc testa diversas variaes de hiperparmetros nesse modelo com todas as combinaes possveis de hiperparmetros em um espao de busca.


Em poucas palavras, hiperprametros se refere a nmeros que voc usa para acertar ou ajustar o desempenho de um modelo estatstico ou de aprendizado de mquina.

Segundo (TOUSHIK) o XGBoost combina principalmente um grande nmero de rvores de regresso com uma pequena taxa de aprendizado. Nesta situao, as rvores adicionadas cedo so significativas e as rvores adicionadas tardiamente no so importantes.Porm, (TOUSHIK) explica uma descoberta importante do mtodo dart, por meio das palavras de **Vinayak e Gilad-Bachrach**. De acordo com eles **propuseram um novo mtodo para adicionar tcnicas de dropout da comunidade de redes neurais profundas a rvores impulsionadas e relataram melhores resultados em algumas situaes,  chamado de dart**. Ele derruba rvores para resolver o **overfitting**. rvores triviais (para corrigir erros triviais) podem ser evitadas. Devido  aleatoriedade introduzida no treinamento, espere algumas diferenas a seguir: O treinamento pode ser mais lento que o gbtree porque o dropout aleatrio impede o uso do buffer de previso. A parada antecipada **(early stopping)** pode no ser estvel, devido  aleatoriedade.

### <center>General parameters</center>

refere-se a qual booster estamos usando para fazer boosting, geralmente rvore ou modelo linear


#### 1. booster [default=gbtree]

- **gbtree** para modelos baseados em rvore; **gblinear** para que os modelos lineares sejam executados em cada iterao.


#### 2. silent [default=0]

- **Ativar o modo silencioso silencioso, defina-o para 1, ou seja, desligado; assim, nenhuma mensagem ser impressa.**Geralmente,  uma boa ideia mant-lo 0, pois as mensagens podem ajudar a entender o modelo; e como as mtricas esto indo.


#### 3. nthread [padro para o nmero mximo de threads disponveis se no for definido]

Isso  usado para processamento paralelo e o nmero de ncleos no sistema deve ser inserido. Se voc deseja executar em todos os ncleos, o valor no deve ser inserido e o algoritmo detectar automaticamente.


### <center>Booster parameters</center>

Ento esta parte descata-se a qual reforo voc escolhar. Temos a seguintes propriedades sobre boosters baseados em rvores destacam-se:

#### 1. eta [default=0.3]

- Semelhante ao learning_rate; trabalha com generalizao.
- Valores finais tpicos a serem usados: 0,01-0,2

Segundo (ANDREW e BRUCE, 2019) o fator de encolhimento ou taxa de aprendizagem est entre 0 e 1 aplicado a  no algoritmo de boosting. O padro  0,3, **mas, para dados ruidosos, recomenda-se valores menores (por exemplo, 0,1)**. Nesse mesmo sentido, (HARRIOSON ,2020) apura que aps cada passo de boosting, (melhoria) os pesos recm-adicionados so escalados de acordo com esse fator. Quanto menor o valor, mais conservador ser, mas tambm sero necessrias mais rvores para convergir. Na chamada ao dados de train, voc pode passar um parmetro learning_rates, que  uma lista de taxas em cada rodada (isto , [.1] * 100 + [.05] * 100)


#### 2. min_child_weight [default=1]

- Define **a soma mnima dos pesos de todas as observaes necessrias em um filho**.  usado para **controlar o ajuste excessivo (overfitting)**. Valores mais altos impedem que um modelo aprenda  relaes que podem ser altamente especficas para a amostra especfica selecionada para uma rvore.


- Valores muito altos podem levar a um ajuste insuficiente, portanto, ele deve ser ajustado usando CV.


#### 3. max_depth [padro=6]

- A **profundidade mxima** de uma rvore tambm  usada para **controlar o ajuste excessivo**, pois uma profundidade maior permitir que o modelo aprenda relaes muito especficas para uma amostra especfica. Ou seja, ao contrrio da floresta aleatria **(random forest)**, que ajusta rvores muito profundas, o **boosting costuma ajustar rvores rasas**. Isso tem a vantagem de evitar interaes falsas e complexas no modelo, as quais podem ser resultado de dados ruidosos (ANDREW e BRUCE, 2019).


- Deve ser ajustado usando CV. Valores tpicos: 3-10


#### 4. max_leaf_nodes

- O nmero mximo de ns terminais ou folhas em uma rvore.

- Como as rvores binrias so criadas, uma profundidade de 'n' produziria um mximo de 2^n folhas, ou seja, em que  cada n da rvore tem  no mximo 2 filhos. Se isso for definido, o GBM ignorar max_depth.


#### 5. gamma [default=0]

- Um n  dividido apenas quando a diviso resultante fornece uma reduo positiva na funo de perda.**Gamma especifica a reduo de perda mnima necessria para que ocorra uma diviso.**

- Torna o algoritmo conservador. Os valores podem variar dependendo da funo de perda e devem ser ajustados. Pode ser usado para controlar o overfitting.

Observa-se (HARRISON, 2020) refina que gama controla a poda (pruning). Assim como, pode varia de 0 a infinito, a reduo de perda mnima necessria para separar mais uma folha. Quanto maior o valor de gama, mais conservador ser. Se as pontuaes de treinamento e de teste divergirem, insira um nmero maior (em torno de 10). Se scores de treinamento e teste estiverem prximas, utilize um nmero menor.

#### 6. max_delta_step [default=0]

- No passo **delta mximo, permitimos que a estimativa de peso de cada rvore seja**. Se o valor for definido como 0, significa que no h restrio. Se estiver definido com um valor positivo, pode ajudar a tornar a etapa de atualizao mais conservadora.

- Isso geralmente no  usado, mas sim, usado para classes desbalanceadas, pois deixa as atualizaes mais conservadoras com valores de 1 a 10.


#### 7. subsample [default=1]

- Denota a frao de observaes a serem amostras aleatrias para cada rvore.

- Valores mais baixos tornam o **algoritmo mais conservador e evitam overfitting**, mas valores muito pequenos podem levar a underfitting.

- Valores tpicos: 0. 5 - 1

- Conforme discutido, a subsample especifica a frao de instncias de treinamento a serem usadas para treinar cada rvore. Por exemplo, se subsample=0.75, cada rvore  treinada em **75% das instncias de treinamento selecionadas aleatoriamente.**Como voc provavelmente j deve ter adivinhado, essa tcnica troca um **vis maior por uma varincia menor.** Ela tambm acelera substancialmente o treinamento. Essa tcnica se chama **gradiente estocstico boosting.**



#### 8. colsample_bytree [padro=1]

- Mostra a frao de colunas a serem amostras aleatrias para cada rvore. Colsample_bytree menor fornece regulao adicional.

- Valores tpicos: 0.5 -1

#### 9. colsample_bylevel [default=1]

- Indica a proporo de **subamostra de colunas para cada diviso, em cada nvel**.

#### 10. colsample_bynode [default=1]

- Indica a proporo de **subamostra de colunas para cada diviso, em cada n**.


#### 11. alpha [default=0]


- **Termo  de regularizao L1.**

- A regularizao L1 fora os pesos dos recursos no informativos a serem zero, subtraindo uma pequena quantidade do peso em cada iterao e, assim, tornando o peso zero, eventualmente. Tambm  chamado de regularizao por esparsidade <a href="https://developers-google-com.translate.goog/machine-learning/crash-course/regularization-for-sparsity/l1-regularization?_x_tr_sl=en&_x_tr_tl=pt&_x_tr_hl=pt-BR&_x_tr_pto=sc">regularizao por esparsidade</a>. Aplica-se a pesos de folha (em vez de pesos de recursos); valor maior significa mais regularizao


#### 12. lambda [default=1]

- **Termo de regularizao L2.**

- Isso costumava lidar com a parte de regularizao do XGBoost. Deve ser explorado para reduzir o overfitting

- A regularizao L2 age como uma fora que remove uma pequena porcentagem de pesos a cada iterao. Portanto, os pesos nunca sero iguais a zero. A regularizao L2 penaliza (weight). Existe um parmetro adicional para ajustar o termo  de regularizao L2 que  chamado de taxa de regularizao <a href="https://developers-google-com.translate.goog/machine-learning/crash-course/regularization-for-simplicity/lambda?_x_tr_sl=en&_x_tr_tl=pt&_x_tr_hl=pt-BR&_x_tr_pto=sc">(lambda)</a>.

- Alm do mais,  mais suave que alfa. Assim sendo, aplica-se aos pesos das folhas.


> *A regularizao  um jeito de evitar o overfitting atravs da incluso de um termo de penalidade no nmero de parmetros (por exemplo, tamanho da rvore) em um modelo.*


#### 12. scale_pos_weight [default=1]

Esse parmetro indica um valor > 0 deve ser usado em casos de problemas relacionados as classes desbalanceadas (classe raras) da classe alta, pois ajuda na convergncia mais rpida.


#### 13. base_score [default=0.5]

- **gain:** o ganho mdio em todas as divises em que o recurso  usado.

- O processo de validao cruzada  ento repetido vrias n rounds vezes, com cada uma das nfold subamostras usadas exatamente uma vez como dados de validao. Todas as observaes so usadas para treinamento e validao.

-  usado para retornar os resultados em um DataFrame de pandas.

**seed**: 

- Este parmetro  usado para gerao de sementes aleatria dos resultados, ou seja, torna as sadas estvel em todas as execues.


```python
# Criando uma funo para facilitar as coisas mais tarde
```

```python
def xgb_model_helper(PARAMETERS,V_PARAM_NAME=False,V_PARAM_VALUES=False,BR=10):
    

    # Cria uma matrix temporria em formato de bit do conjunto de dados a ser treinados
    temp_dmatrix = xgb.DMatrix(data =X_train, label = y_train)
    
    #Check os parmetros a ser utilizados por padro at um certo momento
    if V_PARAM_VALUES==False:
        cv_results = xgb.cv(dtrain=temp_dmatrix, nfold=5,num_boost_round=BR,params=PARAMETERS, as_pandas=True, seed=123)
        return cv_results
    
    else:
        # Criando uma Lista, para armazenar os resultados e os nomes, de cada uma das mtricas.
        results = []
        
        # Percorre a lista de parmetros
        for v_param_value in V_PARAM_VALUES:
            
            # Adicionando o nome dos parmetros avaliado a lista de nomes.
            PARAMETERS[V_PARAM_NAME] = v_param_value
            
            # Treinando o modelo com Cross Validation.
            cv_results = xgb.cv(dtrain= temp_dmatrix, nfold=5, num_boost_round=BR, params=PARAMETERS, as_pandas= True, seed =123)
           
            # Adicionando os resultados gerados a lista de resultados.
            results.append((cv_results["train-auc-mean"].tail().values[-1], cv_results["test-auc-mean"].tail().values[-1]))  # .tail().values[-1] captura somente as colunas
            
        # zip pareia os elementos de uma srie de listas, tuplas ou outras sequncias para criar uma lista de tuplas:

        # Adicionando a mdia da AUC e o desvio-padro dos resultados gerados, pelo modelo analisado ao Dataframe de mdias.
        data = list(zip(V_PARAM_VALUES, results))
        print(pd.DataFrame(data,columns=[V_PARAM_NAME,"auc"]))
        
        return cv_results
```

#### <center>Creating a general  base  model  geral and evaluate performance</center>

```python
#especfica os parmetros desejados 

PARAMETERS={"objective":'binary:logistic', "eval_metric":"auc"}

# chamda da funo
xgb_model_helper(PARAMETERS)
```

### 10.3.2. Otimizao do nmero de rodadas de reforo (j que usamos o DMAtrix do xgb)

```python
# create the DMatrix
fire_dmatrix = xgb.DMatrix(data = X_train, label = y_train)

# Create the parameter dictionary for each tree: params

params = {"objective":"binary:logistic", "max_depth":5}

# Create lis of number of boosting rounds

num_rounds = [5,10,20,25,50,100]

# Empty list to store final round rmse per XGBoost model

final_rmse_per_round = []

# Iterate ove num_rounds and build one model per num_boost_round parameter

for curr_num_rounds in num_rounds:
    
    # Perform cross-validation: cv_results
    cv_results = xgb.cv(dtrain=fire_dmatrix, params= params, nfold=5, num_boost_round = curr_num_rounds, metrics="auc", as_pandas=True, seed= 123)
    
    
    # Append final round RMSE
    final_rmse_per_round.append(cv_results["test-auc-mean"].tail().values[-1])
    
# print the resultant Dataframe

num_rounds_rmses = list(zip(num_rounds,final_rmse_per_round))
print(pd.DataFrame(num_rounds_rmses,columns=["num_boosting_rounds","auc"]))
```

### 10.4.  Ajustando os Parmetros de Regularizao (alfa e lambda)

### 10.4.1. L1 vs L2 Regularizao

De acordo (DEVELOPERS-GOOGLE) esclarece alguns fundamentos especficos sobre as regulariao L2 e L1 que apresentadas da seguintes formas:


L2 e L1 penalizam os pesos de forma diferente:

- L2 penaliza weight.
- L1 penaliza |weight|.

Consequentemente, L2 e L1 tm derivadas diferentes:

- A derivada de L2  **2 * weight**

A derivada de L1  **k (uma constante, cujo valor  independente do peso).**

Nesssa mesma linha de trabalho, (DEVELOPERS-GOOGLE) defende que voc pode pensar na **derivada de L2 como uma fora que remove x% do peso todas as vezes**. DEVELOPERS-GOOGLE ressaltar um conceito importante por meio das palavras do autor **ZENO** de acordo com ele *mesmo se voc remover x por cento de um nmero bilhes de vezes, o nmero diminudo ainda nunca chegar a zero.* (Zeno estava menos familiarizado com as limitaes de preciso de ponto flutuante, que poderiam produzir exatamente zero.) De qualquer forma, **L2 normalmente no leva os pesos a zero.**

Conseqentemente, voc pode pensar na **derivada de L1 como uma fora que subtrai alguma constante do peso todas as vezes**. No entanto, graas aos **valores absolutos, L1 tem uma descontinuidade em 0, o que faz com que os resultados de subtrao que cruzam 0 sejam zerados**. Por exemplo, se a subtrao foraria um peso de +0,1 a -0,2, L 1 definir o peso exatamente 0. <a href ="https://pt.wikipedia.org/wiki/Eureka_(exclama%C3%A7%C3%A3o)">Eureka</a>, L1 zerou o peso.

>A regularizao L 1  penalizando o valor absoluto de todos os pesos, acaba sendo bastante eficiente para modelos largos.

### 10.7.3  Calibration Curve


As curvas de calibrao (tambm conhecidas como diagramas de confiabilidade) comparam quo bem as previses probabilsticas de um classificador binrio so calibradas. Ela plota a frequncia real do **rtulo positivo em relao  probabilidade prevista, para previses agrupadas.** <mark>O eixo x representa a probabilidade mdia prevista em cada classe binria, eixo y  a frao de positivos , ou seja, a proporo de amostras cuja classe  a classe positiva (em cada classe binria)</mark>

A curva de calibrao apresenta uma forma sigmide caracterstica, indicando que o classificador poderia confiar mais em sua intuio e retornar probabilidades mais prximas de 0 ou 1 normalmente.


A calibrao  uma maneira de obter estimativas de probabilidade de qualquer modelo, ou seja, podemos usar a calibrao para garantir que as probabilidades obtidas sejam realmente boas. Portanto, se voc usar a floresta aleatria e us-la para estimar probabilidades, se disser "isso  70% classe um", no est muito claro o que isso significa porque o modelo no  realmente treinado para otimizar estimativas de probabilidade, de modo que as estimativas de probabilidade podem estar erradas bastante.

https://scikit-learn.org/stable/modules/calibration.html

```python
def calib_plot(clf, name, X_test):
    fig = plt.figure(figsize=(6, 6))
    ax1 = plt.subplot2grid(
        (3, 1), (0, 0), rowspan=2
    )
    ax2 = plt.subplot2grid((3, 1), (2, 0))
    ax1.plot(
        [0, 1],
        [0, 1],
        "k:",
        label="Perfectly calibrated",
    )
    clf.fit(X_train, y_train)
    if hasattr(clf, "predict_proba"):
        prob_pos = clf.predict_proba(X_test)[
            :, 1
        ]
    else:  # use decision function
        prob_pos = clf.decision_function(
            X_test
        )
        prob_pos = (
            prob_pos - prob_pos.min()
        ) / (prob_pos.max() - prob_pos.min())
    fraction_of_positives, mean_predicted_value = calibration_curve(
        y_test, prob_pos, n_bins=10
    )
    ax1.plot(
        mean_predicted_value,
        fraction_of_positives,
        "s-",
        label="%s" % (name,),
    )
    ax2.hist(
        prob_pos,
        range=(0, 1),
        bins=10,
        label=name,
        histtype="step",
        lw=2,
    )
    ax1.set_ylabel("Fraction of positives")
    ax1.set_ylim([-0.05, 1.05])
    ax1.legend(loc="lower right")
    ax1.set_title(
        "Calibration plots  (reliability curve)"
    )
    ax2.set_xlabel("Mean predicted value")
    ax2.set_ylabel("Count")
    ax2.legend(loc="upper center", ncol=2)
    plt.tight_layout()
    plt.show()
    return fig
fig = calib_plot(clf_XGB, "XGBoost", X_test)
```


O modelo XGBoost realiza timas previses da classe 0 e 1, ou seja, tendo um pouco mais de previso da classe 1, tal que isso mostrado no histograma inferior que fornece algumas informaes sobre o comportamento do classificador, mostrando o nmero de amostras em cada caixa de probabilidade prevista.

#### O Trade-off do Vis/Varincia


Segundo (GRON) relata um  estudo terico importante da **estatstica e do aprendizado de mquina**   fato de que o erro de generalizao de um modelo pode ser expresso como a soma de trs erros muito diferentes:


    Erro irredutvel
    Erro de varincia
    Erro de vis
> erro = erro irredutvel + vis + varincia


O erroirredutvelourudo, o prprio nome nos diz: no pode ser reduzido. Eleindependntedo algoritmo utilizado e pode ser causado. **A nica maneira de reduzir esta parte do erro  limpar os dados (por exemplo, corrigir as fontes de dados, como sensores quebrados, ou detectar e remover outliers).**


**Varincia**

** a tendncia de se aprender fatos aleatrios independentemente do sinal real.** A varincia  simplesmente o valor esperado dos desvios quadrados de amostragem. **Ela  usado para indicar quo distante, em mdia, o conjunto de estimativas est do valor esperado das estimativas.**A varincia no ajuda a encontrar a preciso total, mas pode encontrar a irregularidade do modelo ao utilizar vrias previses de diferentes conjuntos de dados de treino.

De acordo (GORI, 2017) explica o conceito, que rege os pilares fundamentais da estatstica moderna por meio das palavras **Ronald Fisher em 1925**. De acordo com ele *Fisher desenvolveu o conceito de anlise de varincia, na qual o comportamento de um resultado de interesse pode ser dividido entre aquele devido a fatores controlados e aquele devido a fatores no controlados*. Passo fundamental para esse tipo de anlise foi dado com a publicao do livro Statistical methods for research workers. Para muitos, a obra mais influente da estatstica moderna.

A varincia  prefervel por algumas rezes:

- Pontos negativos e positivos no se cancelam
- Pontos mais distantes recebem peso maior

Um ponto negativo da varincia est no fato da medida no ser calculada na dimenso original dos dados 

- Podemos resolver este problema tomando a raiz da varincia.Pois  este valor  conhecido com **Desvio-padro**.


**Um modelo muito complexo, tem alta varincia por ser capaz de aprender padres que possam no ser reais.**Isso faz com que, de forma contra intuitiva, os modelos mais complexos no sejam sempre a melhor alternativa e isso traz o desafio adicional de buscar o modelo que no seja to complexo, mas ainda assim generalizvel.

> Erro de alta varincia apresentam umagrande flexibilidademaspouca generalizao.Esta parte se deve  sensibilidade excessiva do modelo a pequenas variaes nos dados de treinamento sendo capaz de  causar overfitting


**Vis**

Esta parte do erro de generalizao se deve a suposies erradas, como supor que os dados so lineares quando na verdade so quadrticos.  provvel que um modelo com um vis alto se subajuste (underfitting) aos dados de treinamento

> Vis  a tendncia de o modelo aprender consistentemente uma generalizao incorreta (por exemplo, pela simplicidade do modelo). **O vis  a distncia entre a mdia do conjunto de estimativas e o nico parmetro a ser estimado.**

> O vis estatstico se refere a erros de medio ou amostragem que so  sistemticos e produzidos pelo processo de medio ou amostragem. Deve-se fazer uma distino importante entre erros devido ao acaso e erros devido a vis.

> O vis aparece de diferentes formas e pode ser observvel ou invisvel. Quando um resultado sugere um vis (por exemplo, por referncia a um padro ou a valores reais), costuma ser um indicador de que um modelo estatstico ou de aprendizado de mquina contm erros de especificao, ou que alguma varivel importante foi omitida

```python
# Definindo uma funo, para realizar o clculo do intervalo de confiana, da acurcia e do erro de modelos 
# preditivos de classificao.

def CI(score, nElements,  ci = .95):

    # Define o Z-score, a ser utilizado para o intervalo de confiana selecionado.
    
    z = {
        .90  : 1.645,      # Para Intervalos com 90% de Confiana.
        .95  : 1.96,       # Para Intervalos com 95% de Confiana.
        .98  : 2.326,      # Para Intervalos com 98% de Confiana.
        .99  : 2.576,      # Para Intervalos com 99% de Confiana.
        .995 : 2.807,      # Para Intervalos com 99.5% de Confiana.
        .999 : 3.291       # Para Intervalos com 99.9% de Confiana.
    }

    # Calculando o range de variao do Intervalo.

    interval = z.get(ci) * np.sqrt( (score * (1 - score)) / nElements)
    # retornando do intervalo de confiana.
    return score - interval, score + interval
```

Em poucas palavras, os valores SHAP so usados sempre que voc tem um modelo  complexo (pode ser um GBM, XGBoost, uma Rede Neural ou qualquer coisa que receba alguns  recursos como entrada e produz algumas previses como sada) e voc deseja entender quais

    Os modelos preditivos respondem ao quanto. O SHAP responde ao porqu.
    
    
Os valores SHAP so baseados nos valores Shapley, um conceito vindo da teoria dos jogos. Mas a teoria dos jogos precisa de pelo menos duas coisas: um jogo e alguns jogadores. Como isso 
se aplica  explicabilidade do aprendizado de mquina? Imagine que temos um modelo preditivo, assim temos:


- O jogo  o resultado do modelo.
- Os jogadores so os recursos (variveis) includos no modelo.
    
Assim, o que Shapley faz  quantificar a contribuio que cada jogador traz para o jogo. Ou seja, o que o SHAP faz  quantificar a contribuio que cada recurso traz para a previso feita pelo 
modelo.    

 importante ressaltar que o que chamamos de jogo diz respeito a uma nica  observao. Um jogo = uma observao. De fato, SHAP  sobre interpretabilidade local de um modelo preditivo.

---
#### Como funcionam os valores de Shapley?

O valor de Shapley  a mdia de todas as contribuies marginais para todas as coalizes possveis. **O tempo de computao aumenta exponencialmente com o nmero de recursos. Uma soluo para manter o tempo de computao gerencivel  computar contribuies para apenas algumas amostras das possveis coalizes.**

---
#### Por que usamos o mtodo de valor Shapley?

O valor de Shapley se aplica principalmente em situaes em que as contribuies de cada ator **so** desiguais, mas cada jogador trabalha em cooperao uns com os outros para obter o ganho ou recompensa. O **valor de Shapley** garante que cada ator ganhe tanto ou mais quanto **teria** de agir de forma independente.

---
#### O que  regresso de Shapley?

**A regresso Shapley Value**  uma tcnica para descobrir a importncia relativa das variveis preditoras na regresso linear . Sua principal aplicao  resolver uma fraqueza da **regresso linear**, que  que ela no  confivel quando as variveis previstas so moderadamente a altamente **correlacionadas.**

---
#### Como o Shap  calculado?

A soma dos pesos de todas as contribuies marginais para modelos de 1 caracterstica  deve ser igual  soma dos pesos de todas as contribuies marginais para modelos de 2  caractersticas e assim por diante. Em outras palavras, a soma de todos os pesos na mesma 
linha deve ser igual  soma de todos os pesos em qualquer outra linha.

### 12.3. Modelo

- **Camadas Hide: 4x25**
- **Ativao: Selu**
- **Inicializao: Lecun_normal**
- **Otimizador: Adam**
- **Learning Rate: Performance Scheduling**
- **Regularizao: AlphaDropout**


```python
# Limpamos a sesso Keras e gerando semente aleatria
keras.backend.clear_session()
tf.random.set_seed(42)
np.random.seed(42)

# create model
model = keras.models.Sequential([
    keras.layers.Dense(25,input_dim=6,activation="selu",kernel_initializer="lecun_normal"),
    keras.layers.Dense(25,activation="selu",kernel_initializer="lecun_normal"),
    keras.layers.Dense(25,activation="selu",kernel_initializer="lecun_normal"),
    keras.layers.Dense(25,activation="selu",kernel_initializer="lecun_normal"),
    keras.layers.AlphaDropout(rate=0.2),
    keras.layers.Dense(1, activation="sigmoid")
])

"""
Como estamos fazendo uma classificao binria (com um ou mais rtulos binrios),
usamos o sigmoid (ou seja, a logstica) na funo de ativao na camada de sada
"""

"""
Nesse caso, usaremos a entropia cruzada como argumento de perda. 
Esta perda  para problemas de classificao binria e  definida no Keras como **binary_crossentropy**. 
"""

"""
Compilando o model

Aps a criao de um modelo, devemos chamar seu mtodo compile() para especificar a funo de perda e o otimizador a ser usado. 
Como opo, voc pode especificar uma lista de mtricas extras a serem calculadas durante o treinamento e a avaliao
Ao compilar, devemos especificar algumas propriedades adicionais necessrias ao treinar a rede. 
Lembre-se de treinar uma rede significa encontrar o melhor conjunto de pesos para mapear entradas para sadas em nosso conjunto de dados

"""
# Compilando modelo 
model.compile(loss='binary_crossentropy',
              optimizer = keras.optimizers.Adam(lr=0.003),
               metrics=['accuracy','Precision','Recall','AUC'])

"""
salva os pontos de verificao do modelo em intervalos regulares durante o treinamento, por padro no fim de cada poca:
"""
model_checkpoint_cb = keras.callbacks.ModelCheckpoint("models/fire_model_selu_adam.h5", save_best_only=True)

"""
Uma forma de implementar a early stopping  simplesmente usar a callback da EarlyStopping. Ela interromper o treinamento quando no calcular
nenhum progresso na validao definida para vrias pocas (definida pelo argumento patience) e, opcionalmente, reverter para o melhor modelo. Voc
pode combinar as duas callbacks para salvar os pontos de verificao do modelo (no caso de o computador travar) e interromper o treinamento com
antecedncia, quando no houver mais progresso (a fim de evitar perda de tempo e de recursos)
"""
early_stopping = keras.callbacks.EarlyStopping(patience=20)


#callback TensorBoard()
tensorboard_cb = keras.callbacks.TensorBoard(run_logdir)

# cria um lista de callbacks
callbacks = [early_stopping,model_checkpoint_cb, tensorboard_cb,TqdmCallback(verbose=0)]

# Treina modelo
model.fit(X_train, y_train, epochs=100,batch_size=20,
                              validation_split=0.1,
                              verbose=0,
                              callbacks=callbacks)
# Carrega e Avalia o modelo
model = keras.models.load_model("models/fire_model_selu_adam.h5") # retornar para o melhor modelo
```

### Early stopping ou parada antecipada

Essa tcnica de regularizao interrompe o processo de treinamento assim que o perda de validao atinge um plat (atingir erro mnimo) ou comea a aumentar.Observao, essa  tcnica de reguralio to simples e eficiente que segundo  o cientista da computao **Geoffrey Hinton** chamou de "um lindo almoo grtis".

**Lembra-se, desempenho no conjunto de treinamento for muito melhor do que no conjunto de validao,  provvel que seu modelo esteja tendo overfitting ao conjunto de treinamento (ou existe um bug, como uma incompatibilidade de dados entre o conjunto de treinamento e o conjunto de validao).**


---


O treinamento ocorre ao longo de pocas e cada poca (epoch)  dividida em lotes (batch).

- **Epoch: Uma passagem por todas as linhas no conjunto de dados de treinamento.**
    
- **Lote: Uma ou mais amostras consideradas pelo modelo em uma poca antes que os pesos sejam atualizados.**

**<font color="red"> Uma poca  composta por um ou mais lotes, com base no tamanho de lote escolhido e o modelo  adequado para muitas pocas.</font>**

**Qual  a diferena entre um lote e uma poca em uma rede neural?**

O processo de treinamento ser executado por um nmero fixo de iteraes por meio  do conjunto de dados chamado epochs, que devemos especificar usando o argumento epochs. Tambm devemos definir o nmero de linhas do conjunto de dados que so consideradas antes que os pesos do modelo sejam atualizados em cada poca, chamado de tamanho do lote e definido usando o argumento batch_size.

Essas configuraes podem ser escolhidas experimentalmente por tentativa e erro.Logo, queremos treinar o modelo o suficiente para que ele aprenda um mapeamento bom (ou bom o suficiente) de linhas de dados de entrada para a classificao de sada.
```python
# criando uma funo para calcula as diferena em dados train e test
def diff(x,y):
    
    # criando um lista vazia, para armazena os resultados
    result = []
    
    #loop para percorre dataframe de train
    for i in x :
        #loop para percorre  dataframe de test
        for j in y:
            
            # calcula a diferenas entre os valores com chamada da funo abs
            calcule = abs(i-j)
        
        # add os resultados obtidos na lista
        result.append(calcule)

    return result
```

```python
#Limpamos a sesso Keras e gerando semente aleatria
keras.backend.clear_session()
tf.random.set_seed(42)
np.random.seed(42)

# create model
model_v2 = keras.models.Sequential([
    keras.layers.Dense(100,input_dim=6,activation="selu",kernel_initializer="lecun_normal"),
    keras.layers.Dense(100,activation="selu",kernel_initializer="lecun_normal"),
    keras.layers.Dense(100,activation="selu",kernel_initializer="lecun_normal"),
    keras.layers.Dense(100,activation="selu",kernel_initializer="lecun_normal"),
    keras.layers.Dropout(rate=0.2),
    keras.layers.Dense(1, activation="sigmoid")
])


"""
Nesse caso, usaremos a entropia cruzada como argumento de perda. 
Esta perda  para problemas de classificao binria e  definida no Keras como **binary_crossentropy**. 
"""

# Compilando modelo 
model_v2.compile(loss='binary_crossentropy',
               optimizer = keras.optimizers.Nadam(lr=0.002, beta_1=0.9, beta_2=0.99),
               metrics=['accuracy','Precision','Recall','AUC'])



""""
Na linha do cdigo a seguir, ela calcule o erro de validao a cada N etapas (exatamente como na early stopping) e reduza a taxa de aprendizado por um fator de  quando o erro parar de diminuir.
A callback ReduceLROnPlateau, ela multiplicar a taxa de aprendizado por 0,5  sempre que a melhor perda de validao no melhorar por 5 pocas consecutivas

"""
lr_scheduler = keras.callbacks.ReduceLROnPlateau(factor=0.5, patience=5)


"""
salva os pontos de verificao do modelo em intervalos regulares durante o treinamento, por padro no fim de cada poca:
"""

model_checkpoint_cb = keras.callbacks.ModelCheckpoint("models/fire_model_selu_nadam.h5", save_best_only=True)


# instncia novo log
run_logdir2 = get_run_logdir()

#callback TensorBoard()
tensorboard_cb = keras.callbacks.TensorBoard(run_logdir2)

# cria um lista de callbacks
callbacks = [lr_scheduler,model_checkpoint_cb, tensorboard_cb,TqdmCallback(verbose=0)]

# Treina modelo
model_v2.fit(X_train, y_train, epochs=100,batch_size=32,
                              validation_split=0.1,
                              verbose=0,
                              callbacks=callbacks)

# Carrega e Avalia o modelo
model_v2 = keras.models.load_model("models/fire_model_selu_nadam.h5") # retornar para o melhor modelo
```
