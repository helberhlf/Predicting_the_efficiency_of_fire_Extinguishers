### 1. Definição do problema


De acordo com artigo científico **(Acoustic-Driven Airflow Flame Extinguishing System Design and Analysis of Capabilities of Low Frequency in Different Fuels , 2022)** afirma que os  métodos tradicionais de combate a incêndios colocam em risco os seres humanos e a natureza. Por esta razão, neste estudo (projeto), não foi desenvolvido nenhum sistema de supressão de incêndio de fluxo de ar acionado acústico prejudicial, e experimentos foram realizados para extinguir chamas de gasolina, querosene, combustível líquido diluído (diluente sintético) e gás liquefeito de petróleo (GLP). 17.442 experimentos de extinção foram realizados em 5 tamanhos de chama diferentes, 54 frequências diferentes e em distâncias que variam de 10 cm a 190 cm. Os dados obtidos foram analisados por regressão polinomial. Para combustíveis líquidos, frequência de 10 Hz a 50 Hz, distância de 10 cm a 100 cm, distância de 10 Hz a 32 Hz, distância de 100 cm a 150 cm, frequência de 10 Hz a 28 Hz, distância de 150 cm a 180 cm é a extinção efetiva do incêndio variar. Combustível GLP, 10 a 45 Hz, distância 10 cm a 140 cm, frequência 15 Hz a 30 Hz, distância 140 cm a 180 cm, faixa de extinção eficazes.

Combustível GLP, 10 Hz a 45 Hz a uma distância de 10 cm a 140 cm, frequências de 15 Hz a 30 Hz a uma distância de 140 cm a 180 cm são faixas de extinção eficazes. Além disso, causada pela compressão da membrana dos woofers dentro do colimador e o fluxo de ar injetado na frequência de 30 Hz reduziu a placa metálica de 12 cm de diâmetro de 86,2°C para 18,8°C em 5 min, e a placa metálica deixada esfriar de 80°C a 21,7°C em 10 min a uma distância de 100 cm. O valor médio do Erro Quadrado Médio obtido como resultado da análise de regressão polinomial é 0,9544, e o valor do Erro Quadrado Médio Médio é 1,2020.

### 1.2. Sobre Dataset

O conjunto de dados foi obtido como resultado dos testes de extinção de quatro chamas  de combustíveis diferentes com um sistema de extinção de ondas sonoras. O sistema de extinção  de incêndio por ondas sonoras consiste em 4 subwoofers com uma potência total de 4.000 Watts.  Existem dois amplificadores que permitem que o som chegue a esses subwoofers amplificado. A fonte de alimentação que alimenta o sistema e o circuito do filtro garantindo que  as frequências de som sejam transmitidas adequadamente para o sistema está localizada dentro  da unidade de controle. 

Enquanto o computador é usado como fonte de frequência, o  anemômetro foi usado para medir o fluxo de ar resultante das ondas sonoras durante a fase de  extinção da chama e um decibelímetro para medir a intensidade do som. Um termômetro  infravermelho foi utilizado para medir a temperatura da chama e da lata de combustível, e uma  câmera é instalada para detectar o tempo de extinção da chama. Um total de 17.442 testes foram  realizados com esta configuração experimental. Os experimentos foram planejados da seguinte  forma:

1. Três combustíveis líquidos diferentes e combustível GLP foram usados para criar a chama.
2. 5 tamanhos diferentes de latas de combustível líquido são usados para atingir diferentes tamanhos de chamas.

3. O ajuste de meio e cheio de gás é usado para combustível GLP.

4. Durante a realização de cada experimento, o recipiente de combustível, a 10 cm de distância, foi movido para frente até 190 cm aumentando a distância em 10 cm a cada vez.

5. Junto com o recipiente de combustível, o anemômetro e o decibelímetro foram movidos para frente nas mesmas dimensões.

6. Experimentos de extinção de incêndio foram conduzidos com 54 ondas sonoras de frequências diferentes em cada distância e tamanho de chama.

O conjunto de dados inclui as  características do tamanho do recipiente de combustível representando o tamanho da chama,  tipo de combustível, frequência, decibéis, distância, fluxo de ar e extinção da chama. Assim, 6 
recursos de entrada e 1 recurso de saída serão usados no modelo que você vai construir.  A coluna de status (extinção de chama ou não extinção da chama) pode ser prevista  usando os seis recursos de entrada no conjunto de dados. Os recursos de status e combustível  são categóricos, enquanto outros recursos são numéricos.

### 2. Aprendizagem Supervisionada e Não Supervisionada


Os termos supervisionado e não supervisionado foram herdados do campo de Aprendizado de Máquina, em inglês Machine Learning - ML ou AM. Metaforicamente, um professor "supervisiona “o aluno fornecendo, cuidadosamente, informações alvo, junto com um conjunto de exemplos. Uma tarefa de aprendizado não supervisionada pode implicar no mesmo conjunto de exemplos, mas não incluiria as informações alvo (destino). O aluno não receberia informação sobre o objetivo do aprendizado, mas ficaria livre para tirar as próprias conclusões sobre o que os exemplos têm em comum **(FAWCETT and PROVOST, 2016)**.

Os sistemas de Aprendizado de Máquina podem ser classificados de acordo com a quantidade e o tipo de supervisão que recebem durante o treinamento. Existem quatro categorias principais de aprendizado: supervisionado, não supervisionado, semi-supervisionado e por reforço **(GÉRON, 2021)**
    
    
 ### 2.1. Modelos, Indução e Previsão

Em geral, um modelo é uma representação simplificada da realidade criada para servir a um propósito.Assim,ele se baseia-se em algumas simplificações de suposições sobre o que é importante e o que não é específicos ou às vezes com base em restrições de informações ou tratabilidade. Por exemplo, um mapa é um modelo do mundo físico.Ele abstrai uma enorme quantidade de informações que o cartógrafo considera irrelevantes para sua finalidade. Ele preserva e, às vezes, simplifica ainda mais, as informações relevantes.**(FAWCETT and PROVOST, 2016)**.


Em **Data Science**, um modelo preditivo é uma fórmula para estimar o valor **desconhecido de interesse**: **o alvo**. A fórmula pode ser matemática ou pode ser uma declaração lógica, como uma regra. Muitas vezes, é um híbrido dos dois. Dada nossa divisão de mineração supervisionada de dados em **classificação e regressão**, vamos considerar modelos de classificação (modelos que prevê a estimativa de probabilidade de classe) **(FAWCETT and PROVOST, 2016)**.

<font color='red'>Lembre-se de que os modelos preditivos se concentram em estimar o valor da variável alvo de interesse.</font>

### 2.2. Termo-chave: previsão
No uso comum, **previsão significa antecipar um evento futuro**. Em **Data Science**, de forma mais geral,previsão significa **estimar um valor desconhecido**. Esse valor poderia ser algo no futuro (no uso comum, previsão verdadeira), mas também pode ser algo no presente ou no passado. De fato, como a mineração de dados geralmente lida com dados históricos, os modelos costumam ser construídos e testados por meio do **uso de eventos do passado**. Modelos preditivos para pontuação de crédito estimam a probabilidade de um potencial cliente ser inadimplente. Modelos preditivos para filtragem de spam estimam se determinado e-mail é spam. Modelos preditivos para a detecção de fraude julgam se uma conta foi defraudada. O segredo é que o modelo seja destinado para **estimar  um valor desconhecido**.**(FAWCETT and PROVOST, 2016)**.


### 2.2. Termo-chave: Indução e Dedução

A indução pode ser comparada com a **dedução**.Porém, começa com regras gerais e fatos específicos, e cria outros fatos específicos a partir deles.A utilização dos nossos modelos pode ser considerada um procedimento de dedução (probabilística).**(FAWCETT and PROVOST, 2016)**.

A criação de modelos a partir de dados é conhecida como modelo de indução.A indução é um termo da filosofia que se refere a generalização a partir de casos específicos até regras gerais (ou leis ou verdades). Nossos modelos são regras gerais no sentido estatístico (eles normalmente não se mantêm 100% das vezes; com frequência, nem chegam perto), e o procedimento que cria o modelo a partir de dados é chamado algoritmo de indução ou aprendiz. **(FAWCETT and PROVOST, 2016)**.

```python
def get_size(bytes, suffix="B"):
    """
    Scale bytes to its proper format
    e.g:
        1253656 => '1.20MB'
        1253656678 => '1.17GB'
    """
    factor = 1024
    for unit in ["", "K", "M", "G", "T", "P"]:
        if bytes < factor:
            return f"{bytes:.2f}{unit}{suffix}"
        bytes /= factor
        
# System Information
print("="*40, "System Information", "="*40)
uname = platform.uname()
print(f"System: {uname.system}")
print(f"Release: {uname.release}")
print(f"Version: {uname.version}")
print(f"Machine: {uname.machine}")

# let's print CPU information
print("="*40, "CPU Info", "="*40)
# number of cores
print("Physical cores:", psutil.cpu_count(logical=False))
print("Total cores:", psutil.cpu_count(logical=True))
# CPU frequencies
cpufreq = psutil.cpu_freq()

print(f"Max Frequency: {cpufreq.max:.2f}Mhz")
print(f"Min Frequency: {cpufreq.min:.2f}Mhz")

# Memory Information
print("="*40, "Memory Information", "="*40)

# get the memory details
svmem = psutil.virtual_memory()
print(f"Total: {get_size(svmem.total)}")
```

A maioria dos algoritmos de **Machine Learning** prefere lida com dados números considerando essa suposições; desse modo, converteremos essas categorias de números.

#### 5.4.1.2. Codificando Recursos Categóricos Nominais

- Problema 

Podemos ter recursos (variáveis) com classes nominais que não possui ordenação intrínseca (Blue, Red, Green ).


- Solução 

Codifique o recurso usando o LabelBinarizer do scikit-learn:


Nesse sentido, de acordo (GÉRON , 2021) refina que um dos problemas com essa representação é que os algoritmos de **Machine Learning** assumem que dois valores próximos são mais semelhantes que dois valores
distantes. Isso pode ser bom em alguns casos (por exemplo, para categorias ordenadas como “ruim”, “médio”, “bom” e “excelente”), mas obviamente não é o caso da coluna **fuel** (por exemplo, as categorias 0 e 4 são
mais semelhantes que as categorias 0 e 1). Para corrigir esse problema, uma solução comum é criar um atributo binário (LabelBinarizer) por categoria: um atributo igual a 1 quando a categoria for “<gasoline” (e diferente de 0) e outro atributo igual a 1 quando a categoria for “kerosene” (e diferente de 0) e assim por diante. Isso se chama **codificação one-hot [one-hot encoding ou ainda codificação distribuída]**, porque apenas um atributo será igual a 1, enquanto os outros serão 0. Os atributos novos se chamam de atributos **falsos [dummy]**. 

Segundo (ALBON , 2018 ) ressaltar que a estratégia adequada é criar um recurso binário para cada classe no recurso original. Isso geralmente é chamado de codificação one-hot (na literatura de aprendizado de máquina) ou dummy (na literatura estatística e de pesquisa). O recurso da nossa solução era um vetor contendo quatro classes (ou seja, Gasoline, Kerosene, LPG e  Thinner ). Na codificação **one-hot**, cada classe se torna seu próprio recurso com **1s** quando a classe aparece e **0s** caso contrário. Como nosso recurso tem quatro classes, a **codificação one-hot** retornou quatro recursos binários (um para cada classe). Usando a codificação **one-hot**, podemos capturar a associação de uma observação em uma classe, preservando a noção de que a classe não possui qualquer tipo de hierarquia. Em síntese, vale a pena notar que muitas vezes é recomendado que após uma
codificando um recurso,**descartamos um dos recursos codificados one-hot na matriz resultante para evitar dependência linear.**


Como regra geral, **<font color="red">se o número de categorias for inferior a 10</font>**, a **codificação one-hot** é usualmente a melhor opção (mas seu rendimento pode variar!)

#### Definindo os pipelines

Assim sendo, definimos dois pipelines de pré-processamento de recursos; um para variáveis numéricas (num_pipeline) e outro para variáveis categóricas (cat_pipeline). 

- **num_pipeline** importa a classe  MinMaxScaler() para dimensionamento de dados. 

- **cat_pipeline** importa a classe OneHotEncoder para codificação de dados categóricos como dados numéricos.

- Por fim, combinar  esses dois pipelines usando **ColumnTransformer** para aplicá-los a um conjunto diferentes  colunas.

```python
# construindo um pipeline para pré-processar os atributos numéricos:
num_pipeline = Pipeline([
    ('min_max', MinMaxScaler()),
])

# Criando um  pipeline  para Transformação One Hot Encoding
cat_pipeline = Pipeline([
    ('ohe',OneHotEncoder(handle_unknown='ignore',sparse=False)) #Defina o handle_unknown='ignore' do OneHotEncoder para evitar avisos
],
)
# criando Pipelilne para transformação Label Encoder
le_pipeline = Pipeline([
    ('le', CustomLabelEncoder()), 
])
```


**Extraindo nomes de coluna do ColumnTransformer**


<font color="red">No Scikit-learn a função **ColumnTransformer** é uma ótima ferramenta para pré-processamento de dados, mas retorna um array numpy sem nomes da colunas. Seu método get_feature_names() falhará se pelo menos um transformador não criar novas colunas. Aqui está uma solução rápida para retornar nomes de colunas que funcionam para todos os transformadores e pipelines.</font>


A seguinte função auxiliar é rápida de ser construída em torno do get_feature_names() método do ColumnTransformer, que pode ser encontrado aqui: https://github.com/scikit-learn/scikit-learn/blob/fd237278e895b42abe8d8d09105cbb82dc2cbba7/sklearn/compose/_column_transformer.py#L345

A função percorre as etapas do ColumnTransformere retorna os nomes das colunas de entrada quando o transformador não fornece um get_feature_names() método.Porém, para os pipelines, ele percorre as etapas do pipeline e retornará as colunas de saída do pipeline ou as colunas de entrada, se o pipeline não criar novas colunas.

*References*

<a href ="https://johaupt.github.io/blog/columnTransformer_feature_names.html" target="__blank">Extracting Column Names from the ColumnTransformer</a>

<a href="https://github.com/scikit-learn/scikit-learn/issues/12525">Cannot get feature names after ColumnTransformer #12525</a>

Com base nos conceitos discutido anteriormente, a função adiante é pouco complicada, ela basicamente retorna o nomes das variáveis dos dados transformados ao final da aplicação.O código a seguir é sobre a <a href="https://github.com/scikit-learn/scikit-learn/issues/12525">Cannot get feature names after ColumnTransformer #12525</a>

```python
from skimpy import clean_columns
from sklearn.utils.validation import check_is_fitted

def get_column_names_from_ColumnTransformer(column_transformer, clean_column_names=True, verbose=True):  

    """
    Reference: Kyle Gilde: https://github.com/kylegilde/Kaggle-Notebooks/blob/master/Extracting-and-Plotting-Scikit-Feature-Names-and-Importances/feature_importance.py
    Description: Get the column names from the a ColumnTransformer containing transformers & pipelines
    Parameters
    ----------
    verbose: Bool indicating whether to print summaries. Default set to True.
    Returns
    -------
    a list of the correct feature names
    Note:
    If the ColumnTransformer contains Pipelines and if one of the transformers in the Pipeline is adding completely new columns,
    it must come last in the pipeline. For example, OneHotEncoder, MissingIndicator & SimpleImputer(add_indicator=True) add columns
    to the dataset that didn't exist before, so there should come last in the Pipeline.
    Inspiration: https://github.com/scikit-learn/scikit-learn/issues/12525
    """

    assert isinstance(column_transformer, ColumnTransformer), "Input isn't a ColumnTransformer"
    
    check_is_fitted(column_transformer)

    new_feature_names, transformer_list = [], []

    for i, transformer_item in enumerate(column_transformer.transformers_): 
        transformer_name, transformer, orig_feature_names = transformer_item
        orig_feature_names = list(orig_feature_names)

        if len(orig_feature_names) == 0:
            continue

        if verbose: 
            print(f"\n\n{i}.Transformer/Pipeline: {transformer_name} {transformer.__class__.__name__}\n")
            print(f"\tn_orig_feature_names:{len(orig_feature_names)}")

        if transformer == 'drop':
            continue

        if isinstance(transformer, Pipeline):
            # if pipeline, get the last transformer in the Pipeline
            transformer = transformer.steps[-1][1]

        if hasattr(transformer, 'get_feature_names_out'):
            if 'input_features' in transformer.get_feature_names_out.__code__.co_varnames:
                names = list(transformer.get_feature_names_out(orig_feature_names))
            else:
                names = list(transformer.get_feature_names_out())
        elif hasattr(transformer, 'get_feature_names'):
            if 'input_features' in transformer.get_feature_names.__code__.co_varnames:
                names = list(transformer.get_feature_names(orig_feature_names))
            else:
                names = list(transformer.get_feature_names())

        elif hasattr(transformer,'indicator_') and transformer.add_indicator:
            # is this transformer one of the imputers & did it call the MissingIndicator?

            missing_indicator_indices = transformer.indicator_.features_
            missing_indicators = [orig_feature_names[idx] + '_missing_flag'\
                                  for idx in missing_indicator_indices]
            names = orig_feature_names + missing_indicators

        elif hasattr(transformer,'features_'):
            # is this a MissingIndicator class? 
            missing_indicator_indices = transformer.features_
            missing_indicators = [orig_feature_names[idx] + '_missing_flag'\
                                  for idx in missing_indicator_indices]

        else:

            names = orig_feature_names

        if verbose: 
            print(f"\tn_new_features:{len(names)}")
            print(f"\tnew_features: {names}\n")

        new_feature_names.extend(names)
        transformer_list.extend([transformer_name] * len(names))

    transformer_list, column_transformer_features = transformer_list, new_feature_names

    if clean_column_names:
        new_feature_names = list(clean_columns(pd.DataFrame(columns=new_feature_names)).columns)
    
    return new_feature_names
```

De acordo (GÉRON, 2021) defende que a a detecção de anomalias (também chamada de detecção de outlier) é a tarefa de detectar instâncias que se desviam bastante dos padrões. **Essas instâncias são chamadas de anomalias ou outliers**, ao passo que asinstâncias normais são chamadas de inliers. ### 6.1. Remoção de Outliers usando intervalo interquartil ou IQR via Boxplot

Portanto, antes de entendermos esse método de remoção de valores discrepantes, primeiro precisamos entender os percentis.

### 6.1.2. O que são Percentiles

**Segundo (ANDREW e BRUCE, 2019) uma abordagem diferente para estimar dispersão é com base na observação da distribuição dos dados classificados. A estatística baseada em dados classificados (ordenados) é chamada de estatística ordinal. A medida mais básica é a amplitude: a diferença entre o maior e o menor número. É bom conhecer os valores máximos e mínimos, que são úteis na identificação dos outliers, mas a amplitude é extremamente sensível aos outliers e não é muito útil como medida geral da dispersão dos dados. Para evitar a sensibilidade aos outliers, podemos observar a amplitude dos dados depois da exclusão dos valores de cada ponta. Formalmente, esses tipos de estimativas são baseados em diferenças entre percentis.**<font color="red"> Em um conjunto de dados, o percentil P é um valor em que ao menos P por cento dos valores assume esse valor ou menos, e ao menos (100 – P) por cento dos valores assume esse valor ou mais. Por exemplo, para encontrar o 80° percentil, classifique os dados. Então, começando com o menor valor, avance 80% em direção ao maior valor. Note que a mediana é a mesma coisa que o 50° percentil. O percentil é basicamente o mesmo que um quantil, com quantis indexados por frações (então o quantil .8 é igual ao 80° percentil). Uma medida comum da variabilidade é a diferença entre o 25° percentil e o 75° percentil, chamada de amplitude interquartil (ou IQR).</font>

---
Em outras palavras, um percentil indica o valor abaixo do qual uma determinada porcentagem de observações em um grupo de observações cai.Dessa maneira, pense em classificar o conjunto de dados contendo 100 valores e dividi-lo em 100 partes iguais, agora o valor em qualquer lugar, digamos, no 10º é o nosso 10º percentil, ou seja, o valor no índice 10 indica abaixo do qual 10% dos valores caem.

---
Depois de entender os percentis, é fácil entender o IQR e determinar os limites. Quão? podemos analisar primeiro analisar o IQR.


### 6.1.3.  Interquatile range or IQR

O IQR descreve os 50% dos valores quando classificados do menor para o maior. IQR é a diferença entre o terceiro quartil e o primeiro quartil.

**Assim,o IQR seria  grau confiança ou intervalo de confiança 25% (conhecido como Q1) dos dados estão abaixo dele e o quartil 75% (Q3) é o valor onde 25% dos dadaos estão acima desse. De fato, esse conceito de outliers  via boxplots, foi método mais simples onde os estatísticos da época conseguiria detectar um outlier de forma visual para uma variável e formulou essa fórmula na ilustrado na figura a seguir**


</br><center >Source Image: (TYAGI, 2019)</center></br>
Portanto, pesquisadores vem elaborandos  técnicas mais sofisticadas para detecção de outliers, um ponto importante nesse meio termo é a explosão da **Data Science agredado com Big Data e Estatística moderna**. Atualmente temos os intervalos de confiança frequentemente os mais usados **80%, 85% ,90%, 95% e 99%**


---
**<mark>Portanto, um regra prática é aparar os 10 % iniciais e finais dos dados ofererá , quase sempre, uma proteção contra outliers, exceto nos conjunto de dados menores.</mark>** **(ANDREW e BRUCE, 2019)**.


### 6.1.3. O que é quartis?

Da mesma forma percentil, um quartil corta os dados em 4 partes iguais. Portanto, o percentil 25 é nosso primeiro quartil, o percentil 50 é o segundo quartil, o percentil 75 é nosso terceiro quartil.

Dessa forma  podemos implementa uma função para analisar de acordo aos conceitos discutido antes, preservando 5% dos dados iniciais e finais, ou seja, com **(Q1 = 5% e Q3 = 95%)**, lembrando que devemos sempre começa a detecção com valores conservadores, assim como, o que estiver fora desse intervalo será considerado outliers

```python

'''A função retorna um vetor booleano: True se o elemento for um outlier. Falso, caso contrário.
Agora, para substituir os outliers superiores e inferiores, vamos escrever outra pequena função e aplicá-la em todo o dataframe:'''
def outliers(x): 
       return np.abs(x- x.median()) > 1.5*(x.quantile(0.95)-x.quantile(0.05))
    
'''Substitua os valores outliers mais altos pelo percentil 95 e os valores outliers mais baixos pelo percentil 5'''    
def replace(x):
    out = x[outliers(x)] # é um vector
    return x.replace(to_replace = [out.min(), out.max()],
                    value = (np.percentile(x,5), np.percentile(x,95)))
```

```python
# Aplique replace() em cada coluna do dataframe
df = df.apply(replace,axis=1)
print('Dados antes da aplicação',df.shape)

# Por fim, remova as linhas que contêm qualquer valores outliers pelo método mais simples 
df = df[~df.apply(outliers).any(axis=1)]

print('\nDados depois da aplicação',df.shape)
```

### 7.1 Split 


Nosso objetivo não é avaliar o desempenho do modelo em nossos dados de treinamento, mas o desempenho em dados nunca vistos antes. **Por esse motivo, nosso método de avaliação deve nos ajudar a entender como os modelos são capazes de fazer previsões a partir de dados que nunca viram antes.**


Nesse sentido (ALBON, 2018) refina que uma estratégia pode ser adiar uma fatia de dados para teste. Isso é chamado de validação (ou retenção). Na validação, nossas observações (recursos e alvos) são dividido em dois conjuntos, tradicionalmente chamados de conjunto de **treinamento e conjunto de teste**. Pegamos o conjuntdo de teste e o colocamos de lado, fingindo que nunca o vimos antes. Em seguida, treinamos nosso modelo usando nosso conjunto de treinamento, usando os recursos e o vetor de destino para ensinar o modelo a fazer a melhor previsão. Por fim, simulamos nunca ter visto dados externos, avaliando o desempenho do nosso modelo treinado em nosso conjunto de treinamento em nosso conjunto de teste. **No entanto, a abordagem de validação tem duas grandes fraquezas. Primeiro, o desempenho do modelo pode ser altamente dependente de quais poucas observações foram selecionadas para o conjunto de teste. Segundo, o modelo não está sendo treinado usando todos os dados disponíveis e não está sendo avaliado em todos os dados disponíveis.**


Uma estratégia melhor, que supera essas fraquezas, é chamada de validação cruzada k-fold (KFCV). No KFCV, dividimos os dados em k partes chamadas “folds”. O modelo é então treinado usando k – 1 dobras combinadas em um conjunto de treinamento e então a última dobra é usada como um conjunto de teste. Repetimos isso k vezes, cada vez usando uma dobra diferente como conjunto de teste. O desempenho no modelo para cada uma das k iterações é então calculado para produzir uma medição geral.

Nessa mesma linha de trabalho, (ALBON, 2018) ressaltar que há três pontos importantes a serem considerados quando estamos usando KFCV. Primeiro, KFCV assume que cada observação foi criada independente da outra (ou seja, os dados são distribuídos de forma idêntica independente.Se os dados forem **distribuídos de forma idêntica independente**, é uma boa ideia embaralhar as observações ao atribuir as (“seções/dobras”). No scikit-learn podemos definir **shuffle=True** para realizar o embaralhamento.Em segundo lugar, quando estamos usando KFCV para avaliar um classificador, geralmente é benéfico ter dobras contendo aproximadamente a mesma porcentagem de observações de cada uma das diferentes classes de destino (chamadas k-fold estratificada). <font color ="#f000">Por exemplo, se nosso vetor de destino contiver gênero e 80% das observações forem masculinas, cada dobra conterá 80% observações masculinas e 20% femininas. No Scikit-learn, podemos coordenar a validação cruzada de k-folds estratificadas substituindo a classe KFold por StratifiedKFold.</font>

Dessa forma, se analisarmos a classe target ou alvo do projeto, conforme mostra a figura a seguir, não precisamos  usar **StratifiedKFold** para este propósito, pois a classe destino está relativamente bem distribuída.




**StratifiedKFold**

Se usarmos o KFold para um conjunto de dados desbalanceado, podemos acabar  com dados de treinamento que contém poucas ou nenhuma classe minoritária.Portanto, para evitar esse problema, usamos o StratifiedKFold. **StratifiedKFold é uma versão melhorada de k-Fold que retorna folds estratificadas: cada conjunto contém aproximadamente a mesma porcentagem de amostras de cada classe de destino que o conjunto completo**.

Em outras palavras, StratifiedKFold embaralha seus dados (amostras sem reposição), depois divide os dados em n_splits partes e pronto. Agora, ele usará cada parte como um conjunto de teste. Observe que apenas e sempre embaralha os dados uma vez antes de dividir. Isso significa que cada parte preserva a mesma porcentagem de amostras de cada classe (rótulo) que os dados originais.

> O StratifiedKFold é usado quando é necessário balancear a porcentagem de cada classe para treinar e testar. Se não for necessário KFold é usado.

### 8. Detecção de anomalias usando as misturas de gaussianas


Segundo (GÉRON, 2021) um modelo de **mistura de gaussianas (GMM)** é um modelo probabilístico que assume que as instâncias foram geradas a partir de uma mistura de várias distribuições gaussianas cujos parâmetros são desconhecidos. Todas as instâncias geradas a partir de uma única distribuição gaussiana formam um cluster que normalmente se parece com um elipsoide. Cada cluster pode ter uma forma, tamanho, densidade e orientação elipsoidal diferentes, conforme demonstrado na (Figura adiante). Ao observar uma instância, você sabe que ela foi gerada a partir de uma das distribuições gaussianas, mas não sabe dizer qual delas e não sabe quais são os parâmetros dessas distribuições.



Oberserva-se (GÉRON, 2021) ressaltar que **usar um modelo de mistura de gaussianas para detecção de anomalias é bastante simples: qualquer instância localizada em uma região de baixa densidade pode ser considerada uma anomalia**. Você deve definir o limiar de densidade que quer usar. Por exemplo, em uma empresa de manufatura que tenta identificar produtos defeituosos, quando a proporção desses produtos geralmente é conhecida. Digamos que seja igual a 4%. Logo, você define o limiar de densidade como um valor de 4% resultante das instâncias localizadas em áreas abaixo desse limiar. Caso perceba que obtém muitos falsos positivos (ou seja, produtos perfeitamente bons e sinalizados como defeituosos), você pode diminuir o limiar. Por outro lado, caso tenha muitos falsos negativos (ou seja, produtos defeituosos que o sistema não sinaliza como tal), poderá aumentar esse limiar. Isso se chama trade-off precisão/revocação


> **Os modelos de mistura de gaussianas tentam ajustar todos os dados, incluindo os outliers; desse modo, se você tiver muitos deles, isso influencia a visão de “normalidade” do modelo, e alguns outliers podem ser considerados erroneamente como normais. Se isso ocorrer, você pode tentar ajustar o modelo uma vez, usá-lo para detectar e remover os outliers mais extremos e ajustar o modelo novamente no conjunto de dados limpo. Outra abordagem é utilizar métodos robustos de estimativa de covariância (confira a classe EllipticEnvelope).**

> O que significa ajuste de hiperparâmetro?

Segundo ((TOUSHIK) defende que no aprendizado de máquina, **a otimização ou ajuste de hiperparâmetros é o problema de escolher um conjunto de hiperparâmetros ideais para um algoritmo de aprendizado.** Um hiperparâmetro é um parâmetro cujo valor é usado para controlar o processo de aprendizado, ou seja, é uma configuração externa ao modelo e cujo valor não pode ser estimado a partir de dados. Por outro lado, os valores de outros parâmetros (normalmente os pesos dos nós) são aprendidos. O processo de ajuste de hiperparâmetro é uma caminhada na corda bamba para alcançar um equilíbrio entre **underfitting e overfitting**. Underfitting é quando o modelo de aprendizado de máquina não consegue reduzir o erro para o conjunto de teste ou treinamento.

--- 
> Suas principais características são:

- Eles são frequentemente usados em processos para ajudar a estimar os parâmetros do modelo.

- Eles são frequentemente especificados pelo pesquisador. 

- Eles geralmente podem ser configurados usando heurísticas.

- Eles são frequentemente ajustados para um determinado problema de modelagem preditiva.

---

> O que é exemplo de ajuste de hiperparâmetro?

Observa-se que (TOUSHIK) ressaltar alguns exemplos de hiperparâmetros de modelo incluem: A penalidade na regularização do Classificador de Regressão Logística, ou seja, L1 ou L2. A taxa de aprendizado para treinar uma rede neural. Os hiperpârmetros C e sigma para máquinas vetoriais de suporte.


> Por que usamos o ajuste de hiperparâmetros?

Note-se,(TOUSHIK) refina a importância do ajuste de hiperparâmetros é uma parte essencial do controle do comportamento de um modelo de aprendizado de máquina. Se não ajustarmos corretamente nossos hiperparâmetros, nossos parâmetros de modelo estimados produzirão resultados abaixo do ideal, pois não minimizarão a função de perda. **Isso significa que nosso modelo comete mais erros.**

--- 
Dessa forma, não temos como conhecer o melhor valor para um hiperparâmetro em um determinado problema. Podemos usar regras básicas, copiar valores usados em outros problemas ou  procurar o melhor valor por tentativa e erro. Quando um algoritmo de aprendizagem de máquina é ajustado para um problema  específico, usando por exemplo **grid search ou random search**, você está ajustando os hiperparâmetros do modelo para descobrir os parâmetros do modelo que resultam nas previsões mais precisas.Assim acresce-se que, quando se trata de otimizar um modelo, você testa diversas variações de hiperparâmetros nesse modelo com todas as combinações possíveis de hiperparâmetros em um espaço de busca.


Em poucas palavras, hiperpârametros se refere a números que você usa para acertar ou ajustar o desempenho de um modelo estatístico ou de aprendizado de máquina.

Segundo (TOUSHIK) o XGBoost combina principalmente um grande número de árvores de regressão com uma pequena taxa de aprendizado. Nesta situação, as árvores adicionadas cedo são significativas e as árvores adicionadas tardiamente não são importantes.Porém, (TOUSHIK) explica uma descoberta importante do método dart, por meio das palavras de **Vinayak e Gilad-Bachrach**. De acordo com eles **propuseram um novo método para adicionar técnicas de dropout da comunidade de redes neurais profundas a árvores impulsionadas e relataram melhores resultados em algumas situações, é chamado de dart**. Ele derruba árvores para resolver o **overfitting**. Árvores triviais (para corrigir erros triviais) podem ser evitadas. Devido à aleatoriedade introduzida no treinamento, espere algumas diferenças a seguir: O treinamento pode ser mais lento que o gbtree porque o dropout aleatório impede o uso do buffer de previsão. A parada antecipada **(early stopping)** pode não ser estável, devido à aleatoriedade.

### <center>General parameters</center>

refere-se a qual booster estamos usando para fazer boosting, geralmente árvore ou modelo linear


#### 1. booster [default=gbtree]

- **gbtree** para modelos baseados em árvore; **gblinear** para que os modelos lineares sejam executados em cada iteração.


#### 2. silent [default=0]

- **Ativar o modo silencioso silencioso, defina-o para 1, ou seja, desligado; assim, nenhuma mensagem será impressa.**Geralmente, é uma boa ideia mantê-lo 0, pois as mensagens podem ajudar a entender o modelo; e como as métricas estão indo.


#### 3. nthread [padrão para o número máximo de threads disponíveis se não for definido]

Isso é usado para processamento paralelo e o número de núcleos no sistema deve ser inserido. Se você deseja executar em todos os núcleos, o valor não deve ser inserido e o algoritmo detectará automaticamente.


### <center>Booster parameters</center>

Então esta parte descata-se a qual reforço você escolhará. Temos a seguintes propriedades sobre boosters baseados em árvores destacam-se:

#### 1. eta [default=0.3]

- Semelhante ao learning_rate; trabalha com generalização.
- Valores finais típicos a serem usados: 0,01-0,2

Segundo (ANDREW e BRUCE, 2019) o fator de encolhimento ou taxa de aprendizagem está entre 0 e 1 aplicado a α no algoritmo de boosting. O padrão é 0,3, **mas, para dados ruidosos, recomenda-se valores menores (por exemplo, 0,1)**. Nesse mesmo sentido, (HARRIOSON ,2020) apura que após cada passo de boosting, (melhoria) os pesos recém-adicionados são escalados de acordo com esse fator. Quanto menor o valor, mais conservador será, mas também serão necessárias mais árvores para convergir. Na chamada ao dados de train, você pode passar um parâmetro learning_rates, que é uma lista de taxas em cada rodada (isto é, [.1] * 100 + [.05] * 100)


#### 2. min_child_weight [default=1]

- Define **a soma mínima dos pesos de todas as observações necessárias em um filho**. É usado para **controlar o ajuste excessivo (overfitting)**. Valores mais altos impedem que um modelo aprenda  relações que podem ser altamente específicas para a amostra específica selecionada para uma árvore.


- Valores muito altos podem levar a um ajuste insuficiente, portanto, ele deve ser ajustado usando CV.


#### 3. max_depth [padrão=6]

- A **profundidade máxima** de uma árvore também é usada para **controlar o ajuste excessivo**, pois uma profundidade maior permitirá que o modelo aprenda relações muito específicas para uma amostra específica. Ou seja, ao contrário da floresta aleatória **(random forest)**, que ajusta árvores muito profundas, o **boosting costuma ajustar árvores rasas**. Isso tem a vantagem de evitar interações falsas e complexas no modelo, as quais podem ser resultado de dados ruidosos (ANDREW e BRUCE, 2019).


- Deve ser ajustado usando CV. Valores típicos: 3-10


#### 4. max_leaf_nodes

- O número máximo de nós terminais ou folhas em uma árvore.

- Como as árvores binárias são criadas, uma profundidade de 'n' produziria um máximo de 2^n folhas, ou seja, em que  cada nó da árvore tem  no máximo 2 filhos. Se isso for definido, o GBM ignorará max_depth.


#### 5. gamma [default=0]

- Um nó é dividido apenas quando a divisão resultante fornece uma redução positiva na função de perda.**Gamma especifica a redução de perda mínima necessária para que ocorra uma divisão.**

- Torna o algoritmo conservador. Os valores podem variar dependendo da função de perda e devem ser ajustados. Pode ser usado para controlar o overfitting.

Observa-se (HARRISON, 2020) refina que gama controla a poda (pruning). Assim como, pode varia de 0 a infinito, a redução de perda mínima necessária para separar mais uma folha. Quanto maior o valor de gama, mais conservador será. Se as pontuações de treinamento e de teste divergirem, insira um número maior (em torno de 10). Se scores de treinamento e teste estiverem próximas, utilize um número menor.

#### 6. max_delta_step [default=0]

- No passo **delta máximo, permitimos que a estimativa de peso de cada árvore seja**. Se o valor for definido como 0, significa que não há restrição. Se estiver definido com um valor positivo, pode ajudar a tornar a etapa de atualização mais conservadora.

- Isso geralmente não é usado, mas sim, usado para classes desbalanceadas, pois deixa as atualizações mais conservadoras com valores de 1 a 10.


#### 7. subsample [default=1]

- Denota a fração de observações a serem amostras aleatórias para cada árvore.

- Valores mais baixos tornam o **algoritmo mais conservador e evitam overfitting**, mas valores muito pequenos podem levar a underfitting.

- Valores típicos: 0. 5 - 1

- Conforme discutido, a subsample especifica a fração de instâncias de treinamento a serem usadas para treinar cada árvore. Por exemplo, se subsample=0.75, cada árvore é treinada em **75% das instâncias de treinamento selecionadas aleatoriamente.**Como você provavelmente já deve ter adivinhado, essa técnica troca um **viés maior por uma variância menor.** Ela também acelera substancialmente o treinamento. Essa técnica se chama **gradiente estocástico boosting.**



#### 8. colsample_bytree [padrão=1]

- Mostra a fração de colunas a serem amostras aleatórias para cada árvore. Colsample_bytree menor fornece regulação adicional.

- Valores típicos: 0.5 -1

#### 9. colsample_bylevel [default=1]

- Indica a proporção de **subamostra de colunas para cada divisão, em cada nível**.

#### 10. colsample_bynode [default=1]

- Indica a proporção de **subamostra de colunas para cada divisão, em cada nó**.


#### 11. alpha [default=0]


- **Termo  de regularização L1.**

- A regularização L1 força os pesos dos recursos não informativos a serem zero, subtraindo uma pequena quantidade do peso em cada iteração e, assim, tornando o peso zero, eventualmente. Também é chamado de regularização por esparsidade <a href="https://developers-google-com.translate.goog/machine-learning/crash-course/regularization-for-sparsity/l1-regularization?_x_tr_sl=en&_x_tr_tl=pt&_x_tr_hl=pt-BR&_x_tr_pto=sc">regularização por esparsidade</a>. Aplica-se a pesos de folha (em vez de pesos de recursos); valor maior significa mais regularização


#### 12. lambda [default=1]

- **Termo de regularização L2.**

- Isso costumava lidar com a parte de regularização do XGBoost. Deve ser explorado para reduzir o overfitting

- A regularização L2 age como uma força que remove uma pequena porcentagem de pesos a cada iteração. Portanto, os pesos nunca serão iguais a zero. A regularização L2 penaliza (weight)². Existe um parâmetro adicional para ajustar o termo  de regularização L2 que é chamado de taxa de regularização <a href="https://developers-google-com.translate.goog/machine-learning/crash-course/regularization-for-simplicity/lambda?_x_tr_sl=en&_x_tr_tl=pt&_x_tr_hl=pt-BR&_x_tr_pto=sc">(lambda)</a>.

- Além do mais, é mais suave que alfa. Assim sendo, aplica-se aos pesos das folhas.


> *A regularização é um jeito de evitar o overfitting através da inclusão de um termo de penalidade no número de parâmetros (por exemplo, tamanho da árvore) em um modelo.*


#### 12. scale_pos_weight [default=1]

Esse parâmetro indica um valor > 0 deve ser usado em casos de problemas relacionados as classes desbalanceadas (classe raras) da classe alta, pois ajuda na convergência mais rápida.


#### 13. base_score [default=0.5]

- thersold incial

- A maioria dos classificadores binários que fazem a predição de probabilidades tem um **discrimination threshold** de **50%**. Se a probabilidade prevista estiver acima de **50%**, o classificador atribuirá um rótulo positivo.

- Desde modo, como regra geral, um **threshold** é equilibrado entre os casos é definido de **0,5 ou 50% de probabilidade**. No entanto, esse **threshold** pode não ser o **threshold** ideal: muitas vezes.

####  14. random_state [default=0]

- geração de semente aleatória (seed)

#### 15. importance_type [default='gain']

- **weigh:** o número de vezes que um recurso (variável) é usado para dividir os dados em todas as árvores.

- **gain:** o ganho médio em todas as divisões em que o recurso é usado.

- **cover:** a cobertura média em todas as divisões em que o recurso é usado.

- **total_gain:** o ganho total em todas as divisões em que o recurso é usado.

- **total_cover:** a cobertura total em todas as divisões em que o recurso é usado.


#### 16. missing [default=None]

- Indica o valor de interpretação para missing none que dizer np.nan.

#### 17. n_estimators [default=100]

- o número de árvores a serem geradas, ou seja, número de rodadas ou árvores melhoradas.

- O processo de validação cruzada é então repetido várias n rounds vezes, com cada uma das nfold subamostras usadas exatamente uma vez como dados de validação. Todas as observações são usadas para treinamento e validação.

> Em poucas palavras, o número de seções ou dobras a serem usadas para validação cruzada

**metrics**

- as métricas a serem usadas para avaliar nosso modelo

- É usado para retornar os resultados em um DataFrame de pandas.

**seed**: 

- Este parâmetro é usado para geração de sementes aleatória dos resultados, ou seja, torna as saídas estável em todas as execuções.


```python
# Criando uma função para facilitar as coisas mais tarde
```

```python
def xgb_model_helper(PARAMETERS,V_PARAM_NAME=False,V_PARAM_VALUES=False,BR=10):
    

    # Cria uma matrix temporária em formato de bit do conjunto de dados a ser treinados
    temp_dmatrix = xgb.DMatrix(data =X_train, label = y_train)
    
    #Check os parâmetros a ser utilizados por padrão até um certo momento
    if V_PARAM_VALUES==False:
        cv_results = xgb.cv(dtrain=temp_dmatrix, nfold=5,num_boost_round=BR,params=PARAMETERS, as_pandas=True, seed=123)
        return cv_results
    
    else:
        # Criando uma Lista, para armazenar os resultados e os nomes, de cada uma das métricas.
        results = []
        
        # Percorre a lista de parâmetros
        for v_param_value in V_PARAM_VALUES:
            
            # Adicionando o nome dos parâmetros avaliado a lista de nomes.
            PARAMETERS[V_PARAM_NAME] = v_param_value
            
            # Treinando o modelo com Cross Validation.
            cv_results = xgb.cv(dtrain= temp_dmatrix, nfold=5, num_boost_round=BR, params=PARAMETERS, as_pandas= True, seed =123)
           
            # Adicionando os resultados gerados a lista de resultados.
            results.append((cv_results["train-auc-mean"].tail().values[-1], cv_results["test-auc-mean"].tail().values[-1]))  # .tail().values[-1] captura somente as colunas
            
        # zip “pareia” os elementos de uma série de listas, tuplas ou outras sequências para criar uma lista de tuplas:

        # Adicionando a média da AUC e o desvio-padrão dos resultados gerados, pelo modelo analisado ao Dataframe de médias.
        data = list(zip(V_PARAM_VALUES, results))
        print(pd.DataFrame(data,columns=[V_PARAM_NAME,"auc"]))
        
        return cv_results
```

#### <center>Creating a general  base  model  geral and evaluate performance</center>

```python
#específica os parâmetros desejados 

PARAMETERS={"objective":'binary:logistic', "eval_metric":"auc"}

# chamda da função
xgb_model_helper(PARAMETERS)
```

### 10.3.2. Otimização do número de rodadas de reforço (já que usamos o DMAtrix do xgb)

```python
# create the DMatrix
fire_dmatrix = xgb.DMatrix(data = X_train, label = y_train)

# Create the parameter dictionary for each tree: params

params = {"objective":"binary:logistic", "max_depth":5}

# Create lis of number of boosting rounds

num_rounds = [5,10,20,25,50,100]

# Empty list to store final round rmse per XGBoost model

final_rmse_per_round = []

# Iterate ove num_rounds and build one model per num_boost_round parameter

for curr_num_rounds in num_rounds:
    
    # Perform cross-validation: cv_results
    cv_results = xgb.cv(dtrain=fire_dmatrix, params= params, nfold=5, num_boost_round = curr_num_rounds, metrics="auc", as_pandas=True, seed= 123)
    
    
    # Append final round RMSE
    final_rmse_per_round.append(cv_results["test-auc-mean"].tail().values[-1])
    
# print the resultant Dataframe

num_rounds_rmses = list(zip(num_rounds,final_rmse_per_round))
print(pd.DataFrame(num_rounds_rmses,columns=["num_boosting_rounds","auc"]))
```

### 10.4.  Ajustando os Parâmetros de Regularização (alfa e lambda)

### 10.4.1. L1 vs L2 Regularização

De acordo (DEVELOPERS-GOOGLE) esclarece alguns fundamentos específicos sobre as regulariação L2 e L1 que apresentadas da seguintes formas:


L2 e L1 penalizam os pesos de forma diferente:

- L2 penaliza weight².
- L1 penaliza |weight|.

Consequentemente, L2 e L1 têm derivadas diferentes:

- A derivada de L2 é **2 * weight**

A derivada de L1 é **k (uma constante, cujo valor é independente do peso).**

Nesssa mesma linha de trabalho, (DEVELOPERS-GOOGLE) defende que você pode pensar na **derivada de L2 como uma força que remove x% do peso todas as vezes**. DEVELOPERS-GOOGLE ressaltar um conceito importante por meio das palavras do autor **ZENO** de acordo com ele *mesmo se você remover x por cento de um número bilhões de vezes, o número diminuído ainda nunca chegará a zero.* (Zeno estava menos familiarizado com as limitações de precisão de ponto flutuante, que poderiam produzir exatamente zero.) De qualquer forma, **L2 normalmente não leva os pesos a zero.**

Conseqüentemente, você pode pensar na **derivada de L1 como uma força que subtrai alguma constante do peso todas as vezes**. No entanto, graças aos **valores absolutos, L1 tem uma descontinuidade em 0, o que faz com que os resultados de subtração que cruzam 0 sejam zerados**. Por exemplo, se a subtração forçaria um peso de +0,1 a -0,2, L 1 definirá o peso exatamente 0. <a href ="https://pt.wikipedia.org/wiki/Eureka_(exclama%C3%A7%C3%A3o)">Eureka</a>, L1 zerou o peso.

>A regularização L 1  penalizando o valor absoluto de todos os pesos, acaba sendo bastante eficiente para modelos largos.

### 10.7.3  Calibration Curve


As curvas de calibração (também conhecidas como diagramas de confiabilidade) comparam quão bem as previsões probabilísticas de um classificador binário são calibradas. Ela plota a frequência real do **rótulo positivo em relação à probabilidade prevista, para previsões agrupadas.** <mark>O eixo x representa a probabilidade média prevista em cada classe binária, eixo y é a fração de positivos , ou seja, a proporção de amostras cuja classe é a classe positiva (em cada classe binária)</mark>

A curva de calibração apresenta uma forma sigmóide característica, indicando que o classificador poderia confiar mais em sua “intuição” e retornar probabilidades mais próximas de 0 ou 1 normalmente.


A calibração é uma maneira de obter estimativas de probabilidade de qualquer modelo, ou seja, podemos usar a calibração para garantir que as probabilidades obtidas sejam realmente boas. Portanto, se você usar a floresta aleatória e usá-la para estimar probabilidades, se disser "isso é 70% classe um", não está muito claro o que isso significa porque o modelo não é realmente treinado para otimizar estimativas de probabilidade, de modo que as estimativas de probabilidade podem estar erradas bastante.

https://scikit-learn.org/stable/modules/calibration.html

```python
def calib_plot(clf, name, X_test):
    fig = plt.figure(figsize=(6, 6))
    ax1 = plt.subplot2grid(
        (3, 1), (0, 0), rowspan=2
    )
    ax2 = plt.subplot2grid((3, 1), (2, 0))
    ax1.plot(
        [0, 1],
        [0, 1],
        "k:",
        label="Perfectly calibrated",
    )
    clf.fit(X_train, y_train)
    if hasattr(clf, "predict_proba"):
        prob_pos = clf.predict_proba(X_test)[
            :, 1
        ]
    else:  # use decision function
        prob_pos = clf.decision_function(
            X_test
        )
        prob_pos = (
            prob_pos - prob_pos.min()
        ) / (prob_pos.max() - prob_pos.min())
    fraction_of_positives, mean_predicted_value = calibration_curve(
        y_test, prob_pos, n_bins=10
    )
    ax1.plot(
        mean_predicted_value,
        fraction_of_positives,
        "s-",
        label="%s" % (name,),
    )
    ax2.hist(
        prob_pos,
        range=(0, 1),
        bins=10,
        label=name,
        histtype="step",
        lw=2,
    )
    ax1.set_ylabel("Fraction of positives")
    ax1.set_ylim([-0.05, 1.05])
    ax1.legend(loc="lower right")
    ax1.set_title(
        "Calibration plots  (reliability curve)"
    )
    ax2.set_xlabel("Mean predicted value")
    ax2.set_ylabel("Count")
    ax2.legend(loc="upper center", ncol=2)
    plt.tight_layout()
    plt.show()
    return fig
fig = calib_plot(clf_XGB, "XGBoost", X_test)
```


O modelo XGBoost realiza ótimas previsões da classe 0 e 1, ou seja, tendo um pouco mais de previsão da classe 1, tal que isso mostrado no histograma inferior que fornece algumas informações sobre o comportamento do classificador, mostrando o número de amostras em cada caixa de probabilidade prevista.

#### O Trade-off do Viés/Variância


Segundo (GÉRON) relata um  estudo teórico importante da **estatística e do aprendizado de máquina** é  fato de que o erro de generalização de um modelo pode ser expresso como a soma de três erros muito diferentes:


    Erro irredutível
    Erro de variância
    Erro de viés
> erro = erro irredutível + viés + variância


O erro irredutível ou “ruído”, o próprio nome nos diz: não pode ser reduzido. Ele independnte do algoritmo utilizado e pode ser causado. **A única maneira de reduzir esta parte do erro é limpar os dados (por exemplo, corrigir as fontes de dados, como sensores quebrados, ou detectar e remover outliers).**


**Variância**

**É a tendência de se aprender fatos aleatórios independentemente do sinal real.** A variância é simplesmente o valor esperado dos desvios quadrados de amostragem. **Ela é usado para indicar quão distante, em média, o conjunto de estimativas está do valor esperado das estimativas.**A variância não ajuda a encontrar a precisão total, mas pode encontrar a irregularidade do modelo ao utilizar várias previsões de diferentes conjuntos de dados de treino.

De acordo (GORI, 2017) explica o conceito, que rege os pilares fundamentais da estatística moderna por meio das palavras **Ronald Fisher em 1925**. De acordo com ele *Fisher desenvolveu o conceito de análise de variância, na qual o comportamento de um resultado de interesse pode ser dividido entre aquele devido a fatores controlados e aquele devido a fatores não controlados*. A variância é preferível por algumas rezões:

- Pontos negativos e positivos não se cancelam
- Pontos mais distantes recebem peso maior

Um ponto negativo da variância está no fato da medida não ser calculada na dimensão original dos dados 

- Podemos resolver este problema tomando a raiz da variância.Pois  este valor é conhecido com **Desvio-padrão**.


**Um modelo muito complexo, tem alta variância por ser capaz de aprender padrões que possam não ser reais.**Isso faz com que, de forma contra intuitiva, os modelos mais complexos não sejam sempre a melhor alternativa e isso traz o desafio adicional de buscar o modelo que não seja tão complexo, mas ainda assim generalizável.

> Erro de alta variância apresentam uma grande flexibilidade mas pouca generalização.Esta parte se deve à sensibilidade excessiva do modelo a pequenas variações nos dados de treinamento sendo capaz de  causar overfitting


**Viés**

Esta parte do erro de generalização se deve a suposições erradas, como supor que os dados são lineares quando na verdade são quadráticos. É provável que um modelo com um viés alto se subajuste (underfitting) aos dados de treinamento

> Viés é a tendência de o modelo aprender consistentemente uma generalização incorreta (por exemplo, pela simplicidade do modelo). **O viés é a distância entre a média do conjunto de estimativas e o único parâmetro a ser estimado.**

> O viés estatístico se refere a erros de medição ou amostragem que são  sistemáticos e produzidos pelo processo de medição ou amostragem. Deve-se fazer uma distinção importante entre erros devido ao acaso e erros devido a viés.

> O viés aparece de diferentes formas e pode ser observável ou invisível. Quando um resultado sugere um viés (por exemplo, por referência a um padrão ou a valores reais), costuma ser um indicador de que um modelo estatístico ou de aprendizado de máquina contém erros de especificação, ou que alguma variável importante foi omitida

#### Seleção do Trade-Off entre viés e variância


<center>Source Image: (Data Science Academy)</center>

Assim, um adequado desempenho de um método  aprendizado estatístico em um conjunto de teste **requer baixa variância e baixo viés**.Isso é conhecido de **trade-off**, porque é fácil obter um método com viés  extremamente baixo,mas com alta variância ( por exemplo ,desempenhando uma curva que passa por cada observação de treinamento) ou um método com variância muita baixa, mas com viés alto ( ajustando  uma linha horizontal aos dados).

```python
# Definindo uma função, para realizar o cálculo do intervalo de confiança, da acurácia e do erro de modelos 
# preditivos de classificação.

def CI(score, nElements,  ci = .95):

    # Define o Z-score, a ser utilizado para o intervalo de confiança selecionado.
    
    z = {
        .90  : 1.645,      # Para Intervalos com 90% de Confiança.
        .95  : 1.96,       # Para Intervalos com 95% de Confiança.
        .98  : 2.326,      # Para Intervalos com 98% de Confiança.
        .99  : 2.576,      # Para Intervalos com 99% de Confiança.
        .995 : 2.807,      # Para Intervalos com 99.5% de Confiança.
        .999 : 3.291       # Para Intervalos com 99.9% de Confiança.
    }

    # Calculando o range de variação do Intervalo.

    interval = z.get(ci) * np.sqrt( (score * (1 - score)) / nElements)
    # retornando do intervalo de confiança.
    return score - interval, score + interval
```

### 12.1.2. Métodos agnósticos de modelo local

Os métodos de interpretação local explicam as previsões individuais. Nesta abordagem, iremos aprender sobre os seguintes métodos de explicação local. Desse modo, daremos mais ênfase aos métodos para o problema de negócio.

- Os valores Shapley são um método de atribuição que atribui de forma justa a previsão a recursos individuais.

- SHAP é outro método de cálculo para valores de Shapley, mas também propõe métodos de interpretação global baseados em combinações de valores de Shapley entre os dados.

---

No entato, usaremos a biblioteca SHAP para interpretar o modelo.Essa biblioteca também nos dá insights globais sobre o nosso modelo e ajuda a explicar predições individuais.


**SHAP – que significa SHpley Additive exPlanations – é provavelmente o estado da arte em  explicabilidade de Machine Learning. Esse algoritmo foi publicado pela primeira vez em 2017 por  Lundberg e Lee (abaixo está o link do artigo original) e é uma maneira brilhante de fazer  engenharia reversa da saída de qualquer algoritmo preditivo.**

Link do artigo: https://arxiv.org/abs/1705.07874

Em poucas palavras, os valores SHAP são usados sempre que você tem um modelo  complexo (pode ser um GBM, XGBoost, uma Rede Neural ou qualquer coisa que receba alguns  recursos como entrada e produz algumas previsões como saída) e você deseja entender quais 
decisões o modelo tomou.


    Os modelos preditivos respondem ao “quanto”. O SHAP responde ao “porquê”.
    
    
Os valores SHAP são baseados nos valores Shapley, um conceito vindo da teoria dos jogos. Mas a teoria dos jogos precisa de pelo menos duas coisas: um jogo e alguns jogadores. Como isso 
se aplica à explicabilidade do aprendizado de máquina? Imagine que temos um modelo preditivo, assim temos:


- O “jogo” é o resultado do modelo.
- Os “jogadores” são os recursos (variáveis) incluídos no modelo.
    
Assim, o que Shapley faz é quantificar a contribuição que cada jogador traz para o jogo. Ou seja, o que o SHAP faz é quantificar a contribuição que cada recurso traz para a previsão feita pelo 
modelo.    

É importante ressaltar que o que chamamos de “jogo” diz respeito a uma única  observação. Um jogo = uma observação. De fato, SHAP é sobre interpretabilidade local de um modelo preditivo.

---
#### Como funcionam os valores de Shapley?

O valor de Shapley é a média de todas as contribuições marginais para todas as coalizões possíveis. **O tempo de computação aumenta exponencialmente com o número de recursos. Uma solução para manter o tempo de computação gerenciável é computar contribuições para apenas algumas amostras das possíveis coalizões.**

---
#### Por que usamos o método de valor Shapley?

O valor de Shapley se aplica principalmente em situações em que as contribuições de cada ator **são** desiguais, mas cada jogador trabalha em cooperação uns com os outros para obter o ganho ou recompensa. O **valor de Shapley** garante que cada ator ganhe tanto ou mais quanto **teria** de agir de forma independente.

---
#### O que é regressão de Shapley?

**A regressão Shapley Value** é uma técnica para descobrir a importância relativa das variáveis preditoras na regressão linear . Sua principal aplicação é resolver uma fraqueza da **regressão linear**, que é que ela não é confiável quando as variáveis previstas são moderadamente a altamente **correlacionadas.**

---
#### Como o Shap é calculado?

A soma dos pesos de todas as contribuições marginais para modelos de 1 característica  deve ser igual à soma dos pesos de todas as contribuições marginais para modelos de 2  características e assim por diante. Em outras palavras, a soma de todos os pesos na mesma 
“linha” deve ser igual à soma de todos os pesos em qualquer outra “linha”.

### 12.3. Modelo

- **Camadas Hide: 4x25**
- **Ativação: Selu**
- **Inicialização: Lecun_normal**
- **Otimizador: Adam**
- **Learning Rate: Performance Scheduling**
- **Regularização: AlphaDropout**


```python
# Limpamos a sessão Keras e gerando semente aleatória
keras.backend.clear_session()
tf.random.set_seed(42)
np.random.seed(42)

# create model
model = keras.models.Sequential([
    keras.layers.Dense(25,input_dim=6,activation="selu",kernel_initializer="lecun_normal"),
    keras.layers.Dense(25,activation="selu",kernel_initializer="lecun_normal"),
    keras.layers.Dense(25,activation="selu",kernel_initializer="lecun_normal"),
    keras.layers.Dense(25,activation="selu",kernel_initializer="lecun_normal"),
    keras.layers.AlphaDropout(rate=0.2),
    keras.layers.Dense(1, activation="sigmoid")
])

"""
Como estamos fazendo uma classificação binária (com um ou mais rótulos binários),
usamos o “sigmoid” (ou seja, a logística) na função de ativação na camada de saída
"""

"""
Nesse caso, usaremos a entropia cruzada como argumento de perda. 
Esta perda é para problemas de classificação binária e é definida no Keras como **“binary_crossentropy“**. 
"""

"""
Compilando o model

Após a criação de um modelo, devemos chamar seu método compile() para especificar a função de perda e o otimizador a ser usado. 
Como opção, você pode especificar uma lista de métricas extras a serem calculadas durante o treinamento e a avaliação
Ao compilar, devemos especificar algumas propriedades adicionais necessárias ao treinar a rede. 
Lembre-se de treinar uma rede significa encontrar o melhor conjunto de pesos para mapear entradas para saídas em nosso conjunto de dados

"""
# Compilando modelo 
model.compile(loss='binary_crossentropy',
              optimizer = keras.optimizers.Adam(lr=0.003),
               metrics=['accuracy','Precision','Recall','AUC'])

"""
salva os pontos de verificação do modelo em intervalos regulares durante o treinamento, por padrão no fim de cada época:
"""
model_checkpoint_cb = keras.callbacks.ModelCheckpoint("models/fire_model_selu_adam.h5", save_best_only=True)

"""
Uma forma de implementar a early stopping é simplesmente usar a callback da EarlyStopping. Ela interromperá o treinamento quando não calcular
nenhum progresso na validação definida para várias épocas (definida pelo argumento patience) e, opcionalmente, reverterá para o melhor modelo. Você
pode combinar as duas callbacks para salvar os pontos de verificação do modelo (no caso de o computador travar) e interromper o treinamento com
antecedência, quando não houver mais progresso (a fim de evitar perda de tempo e de recursos)
"""
early_stopping = keras.callbacks.EarlyStopping(patience=20)


#callback TensorBoard()
tensorboard_cb = keras.callbacks.TensorBoard(run_logdir)

# cria um lista de callbacks
callbacks = [early_stopping,model_checkpoint_cb, tensorboard_cb,TqdmCallback(verbose=0)]

# Treina modelo
model.fit(X_train, y_train, epochs=100,batch_size=20,
                              validation_split=0.1,
                              verbose=0,
                              callbacks=callbacks)
# Carrega e Avalia o modelo
model = keras.models.load_model("models/fire_model_selu_adam.h5") # retornar para o melhor modelo
```

> Em vez de passar um conjunto de validação utilizando o argumento validation_data, é possível definir validation_split em relação à proporção do conjunto de treinamento que você deseja que o Keras use para validação. Por exemplo, o validation_split=0.1 informa à biblioteca Keras para usar os últimos 10% dos dados (antes de embaralhá-los) para validação

### Early stopping ou parada antecipada

Essa técnica de regularização interrompe o processo de treinamento assim que o perda de validação atinge um platô (atingir erro mínimo) ou começa a aumentar.Observação, essa é técnica de reguralição tão simples e eficiente que segundo  o cientista da computação **Geoffrey Hinton** chamou de "um lindo almoço grátis".



<center>Image source: (AFSHINE e SHERVINE, 2018)</center>


**Lembra-se, desempenho no conjunto de treinamento for muito melhor do que no conjunto de validação, é provável que seu modelo esteja tendo overfitting ao conjunto de treinamento (ou existe um bug, como uma incompatibilidade de dados entre o conjunto de treinamento e o conjunto de validação).**


---


O treinamento ocorre ao longo de épocas e cada época (epoch) é dividida em lotes (batch).

- **Epoch: Uma passagem por todas as linhas no conjunto de dados de treinamento.**
    
- **Lote: Uma ou mais amostras consideradas pelo modelo em uma época antes que os pesos sejam atualizados.**

**<font color="red"> Uma época é composta por um ou mais lotes, com base no tamanho de lote escolhido e o modelo é adequado para muitas épocas.</font>**

**Qual é a diferença entre um lote e uma época em uma rede neural?**

O processo de treinamento será executado por um número fixo de iterações por meio  do conjunto de dados chamado epochs, que devemos especificar usando o argumento epochs. Também devemos definir o número de linhas do conjunto de dados que são consideradas antes que os pesos do modelo sejam atualizados em cada época, chamado de tamanho do lote e definido usando o argumento batch_size.

Essas configurações podem ser escolhidas experimentalmente por tentativa e erro.Logo, queremos treinar o modelo o suficiente para que ele aprenda um mapeamento bom (ou bom o suficiente) de linhas de dados de entrada para a classificação de saída. O modelo sempre apresentará algum **erro**, mas a quantidade de erro será nivelada após algum ponto para uma determinada configuração do modelo. Isso é chamado de convergência de modelo.

A função que a cada ponto x associa a derivada neste ponto de f(x) é chamada de função derivada de f(x).A figura  seguir simplifica o detalhe do valor máximo e mínimo da função, ou seja, na deridava da função 

a linha azul representa os pontos possíveis de x 

<img src="images/derivada.gif"/>
<center> Source Image: Data Science Academy</center>


Em cada ponto, a derivada de f(x) é a tangente do ângulo que a reta tangente à curva faz em relação ao eixo das abscissas. A reta é sempre tangente à curva azul; a tangente do ângulo que ela faz com o eixo das abscissas é a derivada. Note-se que a derivada é positiva quando verde, negativa quando vermelha, e zero quando preta.

A derivada de uma função y = f(x) num ponto x = x0, é igual ao valor da tangente trigonométrica do ângulo formado pela tangente geométrica à curva representativa de y=f(x), no ponto x = x0, ou seja, a derivada é o coeficiente angular da reta tangente ao gráfico da função no ponto x0.

A função derivada é representada por f'(x).

Criando um  Dataframe para comparação dos Modelos em dados train e test

```python
# criando uma função para calcula as diferença em dados train e test
def diff(x,y):
    
    # criando um lista vazia, para armazena os resultados
    result = []
    
    #loop para percorre dataframe de train
    for i in x :
        #loop para percorre  dataframe de test
        for j in y:
            
            # calcula a diferenças entre os valores com chamada da função abs
            calcule = abs(i-j)
        
        # add os resultados obtidos na lista
        result.append(calcule)

    return result
```

Mas,bem como, podemos observa que os  valores em **teste estam muito próximo ao da validação isso é bom sinal, conforme mostra no dataframe as diferenças são extremamente baixas:) não temos tanto overffiting**

Assim, acrescenta-se que esse modelo atinge AUC em teste **99%** significa que há 99% de chance de que o modelo seja capaz de distinguir entre **classe positiva** "1s" e **classe negativa "0s"**, isso é bom:)

Além disso,quanto maior o valor da AUC, mais eficaz é o classificador. Uma AUC de 1 indica um classificador perfeito: classifica todos os **1s** corretamente, e não classifica erroneamente nenhum **0s** como **1s**

### 12.4. Modelo

- **Camadas Hide: 4x100**
- **Ativação: Selu**
- **Inicialização: lecun_normal**
- **Otimizador: Nadam**
- **Learning Rate: Performance Scheduling**
- **Regularização: AlphaDropout**


```python
#Limpamos a sessão Keras e gerando semente aleatória
keras.backend.clear_session()
tf.random.set_seed(42)
np.random.seed(42)

# create model
model_v2 = keras.models.Sequential([
    keras.layers.Dense(100,input_dim=6,activation="selu",kernel_initializer="lecun_normal"),
    keras.layers.Dense(100,activation="selu",kernel_initializer="lecun_normal"),
    keras.layers.Dense(100,activation="selu",kernel_initializer="lecun_normal"),
    keras.layers.Dense(100,activation="selu",kernel_initializer="lecun_normal"),
    keras.layers.Dropout(rate=0.2),
    keras.layers.Dense(1, activation="sigmoid")
])


"""
Nesse caso, usaremos a entropia cruzada como argumento de perda. 
Esta perda é para problemas de classificação binária e é definida no Keras como **binary_crossentropy**. 
"""

# Compilando modelo 
model_v2.compile(loss='binary_crossentropy',
               optimizer = keras.optimizers.Nadam(lr=0.002, beta_1=0.9, beta_2=0.99),
               metrics=['accuracy','Precision','Recall','AUC'])



""""
Na linha do código a seguir, ela calcule o erro de validação a cada N etapas (exatamente como na early stopping) e reduza a taxa de aprendizado por um fator de λ quando o erro parar de diminuir.
A callback ReduceLROnPlateau, ela multiplicará a taxa de aprendizado por 0,5  sempre que a melhor perda de validação não melhorar por 5 épocas consecutivas

"""
lr_scheduler = keras.callbacks.ReduceLROnPlateau(factor=0.5, patience=5)


"""
salva os pontos de verificação do modelo em intervalos regulares durante o treinamento, por padrão no fim de cada época:
"""

model_checkpoint_cb = keras.callbacks.ModelCheckpoint("models/fire_model_selu_nadam.h5", save_best_only=True)


# instância novo log
run_logdir2 = get_run_logdir()

#callback TensorBoard()
tensorboard_cb = keras.callbacks.TensorBoard(run_logdir2)

# cria um lista de callbacks
callbacks = [lr_scheduler,model_checkpoint_cb, tensorboard_cb,TqdmCallback(verbose=0)]

# Treina modelo
model_v2.fit(X_train, y_train, epochs=100,batch_size=32,
                              validation_split=0.1,
                              verbose=0,
                              callbacks=callbacks)

# Carrega e Avalia o modelo
model_v2 = keras.models.load_model("models/fire_model_selu_nadam.h5") # retornar para o melhor modelo
```
